### 一、简介
　　ConfigParser模块在python3中修改为configparser.这个模块定义了一个ConfigParser类，该类的作用是使用配置文件生效，配置文件的格式和windows的INI文件的格式相同。<br>
　　使用模块中的RawConfigParser()、ConfigParser()、 SafeConfigParser()这三个方法，创建一个对象使用对象的方法对指定的配置文件做增删改查操作。<br>
　　INI配置文件是由小节组成的，每个小节都有一个 [section] 标头，加上多个由特定字符串 (默认为 = 或 : ) 分隔的键/值条目。<br>
>　　1、默认情况下小节名对大小写敏感而键对大小写不敏感。 <br>
　　2、键和值开头和末尾的空格会被移除。<br> 
　　3、值可以被省略，在此情况下键/值分隔符也可以被省略。<br>
　　4、值还可以跨越多行，只要其他行带有比值的第一行更深的缩进。 <br>
　　5、依据解析器的具体模式，空白行可能被视为多行值的组成部分也可能被忽略。<br>
  
　　配置文件可以包含注释，但要带有指定字符前缀 (默认为 # 和 ; )。 注释可以单独出现于原本的空白行，并可使用缩进。<br> 
　　INI配置文件与Linux中repo文件中的格式类似。<br>
格式：<br>
>\[section\] <br>
name=value<br>
或者<br>
name: value<br>
"#" 和";" 表示注释<br>
<br>
\[DEFAULT\] #设置默认的变量值，初始化<br>
\[My Section\]<br>
foodir: %(dir)s/whatever<br>
dir=frob<br>
long: this value continues<br>
   in the next line<br>

%(dir)s 会被frob代替。默认值会以字典的形式传递给ConfigParser的构造器。section一般存放的内置目录下，如果切换到其他的目录需指定存放位置。<br>

### 二、方法
　　注意：在调用这三个函数时，切记函数都会调用optionxform()，在传递键值对数据时，会将键名全部转化为小写。

　　**RawConfigParser()**<br>
> ConfigParser.RawConfigParser(\[defaults\[, dict_type\[, allow_no_value\]\]\]) <br>

　　defaults : 如果指定默认值，则使用默认值的键值对<br>
　　dict_type：使用新的section的键值对<br>
　　allow_no_value ：默认是False，如果是True，表示可以接收空值（None）<br>
　　return：对象<br>
不支持可变参数，在section中不能存在%()s<br>

　　**ConfigParser()**<br>
>ConfigParser.ConfigParser(\[defaults\[, dict_type\[, allow_no_value\]\]\]) <br>

在default中必须出现%()s<br>

　　**SafeConfigParser()**<br>
>ConfigParser.SafeConfigParser(\[defaults\[, dict_type\[, allow_no_value\]\]\]) <br>

更加智能化，在section中是否存在%()s会自动判断<br>

传递参数使用函数optionxform()，foo %(bar)s 和 foo %(BAR)s是相同的，optionxform()会将大写字母全部转换为小写。<br>

　　**常见异常**<br>

|异常	|描述|
|:---|:---|
|ConfigParser.Error	|所有异常的基类|
|ConfigParser.NoSectionError	|指定的section没有找到|
|ConfigParser.DuplicateSectionError	|调用add_section() 时，section名称已经被使用|
|ConfigParser.NoOptionError	|指定的参数没有找到|
|ConfigParser.InterpolationError	|当执行字符串插值时出现问题时，出现异常的基类|
|ConfigParser.InterpolationDepthError	|当字符串插值无法完成时，因为迭代次数超过了最大的范围，所以无法完成。InterpolationError的子类|
|InterpolationMissingOptionError	|当引用的选项不存在时，会出现异常。InterpolationError的子类|
|ConfigParser.InterpolationSyntaxError	|当产生替换的源文本不符合所需的语法时，就会出现异常。InterpolationError的子类。|
|ConfigParser.MissingSectionHeaderError	|当试图解析一个没有分段标题的文件时，会出现异常。|
|ConfigParser.ParsingError	|当试图解析文件时发生错误时，会出现异常|
|ConfigParser.MAX_INTERPOLATION_DEPTH	|当raw参数为false时，get()的递归插值的最大深度。这只适用于ConfigParser类|

### 三、RawConfigParser 对象
　　对象的操作可以分为两大类，一种是对配置文件的操作，另一种是对读取后数据流的操作。

####  1、操作配置文件

　　**读取配置文件**

|方法	|描述|
|:---|:---|
|read(filenames)	|filesnames是一个列表，需要从文件加载初始值的应用程序应该在调用read()之前使用readfp()加载所需的文件或文件。|
|readfp(fp[, filename])	|在fp中，从文件或文件类对象中读取和解析配置数据(只使用readline()方法)。如果文件名被省略，并且fp有一个name属性，它被用于文件名;默认值为\< ? \>。|

　　**写入配置文件**

|方法	|描述|
|:---|:---|
|write(fileobject)	|将配置的表示写入指定的文件对象。这个表示可以由未来的read()调用解析。|

####  2、对内存中数据流的操作

　　**获取所有的section节点**

　　**获取指定section节点的设置**

|方法	|描述|

|options()	|返回指定section节点的设置：名称key（列表）|


　　**增加配置文件中的值**

|方法	|描述|
|:---|:---|

　　**删除配置文件中的值**

|方法	|描述|
|:---|:---|
|remove_option(section, option)	|从指定的部分中删除指定的选项。如果该部分不存在，请提出NoSectionError。如果存在的选项被删除，返回True;否则返回False。|
|remove_section(section)	|从配置中删除指定的section。如果这个部分确实存在，返回True。否则返回假|

　　**修改配置文件中的值**

|方法	|描述|
|:---|:---|
|set(section, option, value)	|如果给定的部分存在，将给定的选项设置为指定的值|

　　**查找配置文件中的值**

|方法	|描述|

### 四、ConfigParser对象
　　ConfigParser中包含RawConfigParser相同的方法（继承），还有一部分增加的方法

|方法	|描述|
|:---|:---|
|get(section, option[, raw[, vars]])	|为指定的section获取一个选项值。如果提供了vars，它必须是一个字典。该选项在vars(如果提供)、分段和默认值中查找,|
|items(section[, raw[, vars]])	|返回给定section中每个选项的(名称、值)对的列表|

　　所有的“%”插值都在返回值中展开，除非原始的参数是真的。内插键的值与选项相同

实例一：写配置文件<br>
> 1 import ConfigParser<br>
 2 <br>
 3 config = ConfigParser.RawConfigParser()<br>
 4 <br>
 5 # When adding sections or items, add them in the reverse order of<br>
 6 # how you want them to be displayed in the actual file.<br>
 7 # In addition, please note that using RawConfigParser's and the raw<br>
 8 # mode of ConfigParser's respective set functions, you can assign<br>
 9 # non-string values to keys internally, but will receive an error<br>
10 # when attempting to write to a file or when you get it in non-raw<br>
11 # mode. SafeConfigParser does not allow such assignments to take place.<br>
12 config.add_section('Section1')<br>
13 config.set('Section1', 'an_int', '15')<br>
14 config.set('Section1', 'a_bool', 'true')<br>
15 config.set('Section1', 'a_float', '3.1415')<br>
16 config.set('Section1', 'baz', 'fun')<br>
17 config.set('Section1', 'bar', 'Python')<br>
18 config.set('Section1', 'foo', '%(bar)s is %(baz)s!')<br>
19 <br>
20 # Writing our configuration file to 'example.cfg'<br>
21 with open('example.cfg', 'wb') as configfile:<br>
22     config.write(configfile)<br>

实例二：读配置文件
>1 import ConfigParser<br>

实例四：默认值
　　所有三种类型的config分析器都可以使用默认值。如果在其他地方没有定义一个选项，那么它们就被用于插值
>1 import ConfigParser<br>
 2 <br>
 3 # New instance with 'bar' and 'baz' defaulting to 'Life' and 'hard' each<br>
 4 config = ConfigParser.SafeConfigParser({'bar': 'Life', 'baz': 'hard'})<br>
 5 config.read('example.cfg')<br>
 6 <br>
 7 print config.get('Section1', 'foo')  # -> "Python is fun!"<br>
 8 config.remove_option('Section1', 'bar')<br>
 9 config.remove_option('Section1', 'baz')<br>
10 print config.get('Section1', 'foo')  # -> "Life is hard!"<br>

实例五：在各section之间移动选项
><br>1 def opt_move(config, section1, section2, option):<br>
2     try:<br>
3         config.set(section2, option, config.get(section1, option, 1))<br>
4     except ConfigParser.NoSectionError:<br>
5         # Create non-existent section<br>
6         config.add_section(section2)<br>
7         opt_move(config, section1, section2, option)<br>
8     else:<br>
9         config.remove_option(section1, option)<br>

实例六：配置文件中有空值
　　一些配置文件包含了没有值的设置，但是它与ConfigParser支持的语法相一致。对构造函数的不允许值参数可以被用来表示应该接受这样的值
><br>
>>> import ConfigParser<br>
>>> import io<br>
<br>
>>> sample_config = """<br>
... [mysqld]<br>
... user = mysql<br>
... pid-file = /var/run/mysqld/mysqld.pid<br>
... skip-external-locking<br>
... old_passwords = 1<br>
... skip-bdb<br>
... skip-innodb<br>
... """<br>
>>> config = ConfigParser.RawConfigParser(allow_no_value=True)<br>
>>> config.readfp(io.BytesIO(sample_config))<br>
<br>
>>> # Settings with values are treated as before:<br>
>>> config.get("mysqld", "user")<br>
'mysql'<br>
<br>
>>> # Settings without values provide None:<br>
>>> config.get("mysqld", "skip-bdb")<br>
<br>
>>> # Settings which aren't specified still raise an error:<br>
>>> config.get("mysqld", "does-not-exist")<br>
Traceback (most recent call last):<br>
  ...<br>
ConfigParser.NoOptionError: No option 'does-not-exist' in section: 'mysqld'<br>
