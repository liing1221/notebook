```python
import numpy as np
a = np.array([0, 1, 2, 3, 4])
b = np.arange(5)
c = np.linspace(0, 2 * np.pi, 8)
print(a, b, c, sep='\n')
print(a[0], a[-2])
print(c.reshape((2,4)))
print(c[[0, 1]])
```

```python
import matplotlib.pyplot as plt
# %matplotlib inline
plt.figure()
a = np.linspace(0, 2 * np.pi, 50)
b = np.sin(a)
plt.plot(a, b)
mask = b >= 0
plt.plot(a[mask], b[mask], 'bo')
mask = (b >= 0) & (a <= np.pi / 2)
plt.plot(a[mask], b[mask], 'go')
plt.show()
```
