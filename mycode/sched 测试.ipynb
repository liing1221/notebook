```python
import time, sched
import datetime


s = sched.scheduler(time.time, time.sleep)
i = 0

def event_fun1():
    print("func1 Time:", datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
    time.sleep(10)


def perform1(inc):
    global i
    i += 1
    print(i,perform1.__name__, flush=True)
    print(s.queue, flush=True)
    s.enter(inc, 0, perform1, (inc,))
    event_fun1()


def event_fun2():
    print("func2 Time:", datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'))


def perform2(inc):
    global i
    i += 1
    print(i,perform2.__name__, flush=True)
    s.enter(inc, 0, perform2, (inc,))
    # print(s.queue, flush=True)
    event_fun2()


def mymain(func, inc=2):
    global i
    if func == "1":
        i += 1
        print(i,mymain.__name__, flush=True)
        s.enter(0, 0, perform1, (3,))# 每隔10秒执行一次perform1
    if func == "2":
        i += 1
        print(i,mymain.__name__, flush=True)
        s.enter(0, 0, perform2, (3,))# 每隔20秒执行一次perform2
if __name__ == '__main__':
    mymain('1')
    mymain('2')
    s.run()
```
