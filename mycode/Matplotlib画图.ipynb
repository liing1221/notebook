## Matplotlib画图

参考：[Matplotlib教程](https://github.com/rougier/matplotlib-tutorial)　[官方教程](https://matplotlib.org/users/pyplot_tutorial.html)

```python
import numpy as np

# 获取正弦余弦函数
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)    #  [-pi, pi]间256个元素的数组
C, S = np.cos(x), np.sin(x)
```

[线属性(Controlling line properties)](https://matplotlib.org/users/pyplot_tutorial.html#controlling-line-properties)　[线API](https://matplotlib.org/api/artist_api.html#matplotlib.lines.Line2D)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(2, 1.5), dpi=100)                            #  创建一个大小为 2*1.5 点的新图形，使用每英寸100个点
plt.subplot(111)                                                # 从1x1的网格创建一个新的子图：111表示纵1图，行1图，放在第1图

plt.plot(x, C, color="blue", linewidth=1.0, linestyle="-")    # 使用宽度为1(像素)的连续线的蓝色绘制余弦曲线
plt.plot(x, S, color="green", linewidth=1.0, linestyle="-")  # 线色、线宽、线形


plt.xlim(-4.0*1.1,4.0*1.1)                                 # 设定x轴范围、界限　　　       limits = [min, max]
plt.ylim(-1.0*1.1,1.0*1.1)                                # 设定y轴范围、界限

plt.xticks(np.linspace(-4,4,9,endpoint=True))          # 设定x标注点  (位置、标注值)　　　ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))         # 设定ｙ标注点
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])  # 分别设置标注点与标注值

# Save figure using 72 dots per inch
# savefig("../figures/exercice_2.png",dpi=72)　保存图片，使用７２像素／英寸

# Show result on screen
plt1 = plt
plt1.show()
```
