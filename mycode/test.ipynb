```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(2, 1, 10000)

# 计算数据的概率密度函数
mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# 绘制数据的直方图和概率密度函数
plt.hist(data, bins=50, density=True, alpha=0.6)
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Probability density')
plt.title('Histogram and Probability Density of Data')
plt.show()

# 计算概率密度函数的极值点
max_pdf_value = p.max()
max_pdf_index = p.argmax()
max_pdf_point = x[max_pdf_index]
print('Maximum PDF value:', max_pdf_value)
print('Maximum PDF point:', max_pdf_point)
```
