## Pycharm 常用快捷键<br>
**重要参考：**(CSDN大神原创)[最全Pycharm教程](https://blog.csdn.net/u013088062/article/details/50100121)<br>
　　　　　　[Pycharm远程连接编程](https://www.cnblogs.com/AlvinSui/p/8081927.html)<br>
　　　　　　[Pycharm2020最新激活码汇总|pycharm永久激活](https://www.jianshu.com/p/16614b6ee4f8)<br>
#### 1、常用快捷键<br>
>Ctrl + Q	快速查看文档<br>
Ctrl + F1	显示错误描述或警告信息<br>
Ctrl + /	行注释（可选中多行）<br>
Ctrl + Alt + L	代码格式化<br>
Ctrl + Alt + O	自动导入<br>
Ctrl + Alt + I	自动缩进<br>
Tab / Shift + Tab	缩进、不缩进当前行（可选中多行）<br>
Ctrl+C/Ctrl+Insert	复制当前行或选定的代码块到剪贴板<br>
Ctrl + D	复制选定的区域或行到后面或下一行<br>
Ctrl + Y	删除当前行<br>
Shift + Enter	下一行另起一行<br>
Ctrl +Ｊ	插入模版<br>
Ctrl + Shift +/-	展开/折叠全部代码块<br>
Ctrl + Numpad+	全部展开<br>
Ctrl + Numpad-	全部折叠<br>
Ctrl + Delete	删除到字符结束<br>
Ctrl + Backspace	删除到字符开始<br>
Ctrl + Shift + F7	将当前单词在整个文件中高亮，F3移动到下一个，ESC取消高亮。<br>
Alt + up/down	方法上移或下移动<br>
Alt + Shift + up/down	当前行上移或下移动<br>
Ctrl + B/鼠标左键	转到方法定义处<br>
Ctrl + W	选中增加的代码块<br>
Shift + F6	方法或变量重命名<br>
Ctrl + E	最近访问的文件<br>
Esc	从其他窗口回到编辑窗口<br>
Shift + Esc	隐藏当前窗口，焦点到编辑窗口<br>
F12	回到先前的工具窗口<br>
