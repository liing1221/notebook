### 一、Psutil模块简介  
　　服务器的监控通过安装一些常用的监控软件之外，有时也需要运行一些shell或Python脚本；shell下可以使用系统自带的ps/free/top/df等shell命令，Python可以调用subprocess等模块来运行shell命令，不过这么做就比较麻烦。这里有一个比较好用的第三方模块：psutil。<br>
　　psutil是一个跨平台的库，用于在Python中检索有关运行进程和系统利用率（CPU，内存，磁盘，网络，传感器）的信息。它主要用于系统监视，分析，限制进程资源和运行进程的管理。它实现了UNIX命令行工具提供的许多功能，例如：ps，top，lsof，netstat，ifconfig，who，df，kill，free，nice，ionice，iostat，iotop，uptime，pidof，tty，taskset，pmap。 <br>
　　psutil目前目前支持Linux、Windows、OS X、FreeBSD和Sun Solaris等装有Python2.6至3.6的32-bit和64-bit架构. 也可以在PyPy上运行。<br>

### 三、Psutil获取CPU信息
使用下面这些方法可以获取计算机CPU数量，CPU使用率：<br>

|方法	|说明|
|:---|:---|
|cpu_count(logical=True)	|获取CPU数量(参数：物理False、逻辑核心数量)|
|cpu_times(percpu=False)	|获取不同状态下CPU运行时间|
|cpu_stats() |CPU统计信息|
|cpu_percent(interval=None, percpu=False)	|获取CPU使用率|
|cpu_times_percent(interval=None, percpu=False)	|获取CPU各状态占百分比|

两个个知识点需要简单介绍下，这里只需要了解即可：<br>
　　1）CPU数量：CPU分为物理CPU与逻辑CPU，之间关系如下：<br>
>1>物理CPU：计算机实际CPU数量，一般个人电脑1个；<br>
2>核数：现在CPU基本都是多核，比如双核，4核；<br>
3>逻辑CPU：物理CPU*核数*超线程数<br>

　　2）CPU运行状态分为用户态与内核态；<br>
>1>用户态是指程序在内存运行，不直接访问硬件资源；<br>
2>用户需要访问硬件资源，需要系统调用，这部分有操作系统（内核）完成，称为内核态，例如：读写文件，网络数据收发，键盘事件获取等；<br>

下面我们来实际操作下（windows环境）：<br>

### 四、Psutil获取内存信息
1.3 获取系统内存信息<br>
程序在内存中运行，实际工作中我们需要对内存进行监控，如果内存使用接近100%，可能存在内存泄漏等问题；我们主要关注内存的total(内存总数)，used(已使用)，free(未使用)情况，相关方法如下：<br>

|方法	|说明|
|:---|:---|
|psutil.virtual_memory	|获取内存使用情况，不同系统返回值不同|
|psutil.swap_memory()	|获取swap（交换）内存使用情况|

实际操作如下（windows环境下）：<br>

>import psutil<br>
mem = psutil.virtual_memory()<br>
print('系统内存：', mem)<br>
print('总物理内存：', mem.total)<br>
print('空闲内存：', mem.available)<br>
print('使用内存：', mem.used)<br>
print('未使用内存：', mem.free)<br>
print('内存使用率：', mem.percent)<br>
print('swap 内存：', psutil.swap_memory())<br>

输出结果：<br>

### 七、Psutil获取进程信息
　　计算机中，每个程序都是一个进程或者多个进程，除去系统占用资源其他都被这些进程占用，比如我们的web服务，数据库等；很多情况下因为程序自身问题，会导致CPU运行100%，内存耗尽，磁盘写满，最后导致服务崩溃，我们可以通过psutil库获取进程相关信息，主要方法如下：<br>

|方法	|说明|
|:---|:---|
|psutil.pids()	|获取进程ID（每个进程都有唯一ID）|
|psutil.Process(pid)	|根据进程ID获取进程Process对象|
|psutil.Popen()|Popen类，获取用户启动的应用程序进程信息，并可跟踪程序进程的运行状态。|

实际操作：<br>

>import psutil<br>
pids = psutil.pids()　　　　# 获取当前所有进程，返回pid组成的列表<br>
print(pids)<br>
process = psutil.Process(pids[0])　　　　　# 获取ID为pids[0]的进程,<br>
print(process)<br>

输出结果：<br>
>[0, 4, 120, 440, 624, 736, 744, ... ...]<br>
psutil.Process(pid=0, name='System Idle Process', started='19:55:14')<br>

结果中可以看到，进程的ID为0， 名称为'System Idle Process'；<br><br>
