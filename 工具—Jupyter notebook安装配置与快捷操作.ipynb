### 一、简介
　　Jupyter Notebook是一个开源的Web应用程序，允许用户创建和共享包含代码、方程式、可视化和文本的文档。它的用途包括：数据清理和转换、数值模拟、统计建模、数据可视化、机器学习等等。它具有以下优势：<br>
　　可选择语言：支持超过40种编程语言，包括Python、R、Julia、Scala等。<br>
　　分享笔记本：可以使用电子邮件、Dropbox、GitHub和Jupyter Notebook Viewer与他人共享。<br>
　　交互式输出：代码可以生成丰富的交互式输出，包括HTML、图像、视频、LaTeX等等。<br>
　　大数据整合：通过Python、R、Scala编程语言使用Apache Spark等大数据框架工具。支持使用pandas、scikit-learn、ggplot2、TensorFlow来探索同一份数据。<br>
### 二、安装与运行
#### 1、安装Jupyter NoteBook
　　有两种安装方式：使用Anaconda安装或使用pip命令安装。关于安装的全部信息可以在官网读到：安装Jupyter。<br>
　　1、Anaconda安装：安装Anaconda，即可自动安装包括Python、Jupyter Notebook和其他常用的科学计算和数据科学软件包。<br>
　　2、Pip安装：<br>
>python3 -m pip install --upgrade pip<br>
pi3 install jupyter<br>

　　成功安装Jupyter Notebook后，在命令行运行以下命令就可打开Jupyter Notebook。<br>
> jupyter notebook<br>
#### 2、配置Jupyter notebook服务密码
　　Jupyter notebook 服务密码注册（也可用于忘记密码的重设）：<br>
　　修改jupyter notebook配置文件jupyter_notebook_config.py，使服务密码为固定密码，不在变化（c.NotebookApp.allow_password_change = False）。然后命令行执行如下命令设置固定密码：<br>
>Jupyter notebook password        # 设置密码
#### 3、更换Python内核：<br>
##### 3.1、找到upyter Notebook内核指定的Python环境位置：<br>
　　在cmd下输入命令：<br>
>ipython kernelspec list<br>

　　输出的结果会是一个位置：<br>
>Available kernels:<br>
　　python3　　C:\Liing\Anaconda3\share\jupyter\kernels\python3<br>
    
　　在文件资源管理器中打开这个文件夹，文件夹里会有几个文件，jupyter notebook 即使用kernel.json文件配置内核环境。<br>
##### 3.2、修改Python环境位置为指定位置即可：<br>
　　使用文本编辑器打开kernel.json文件，里面的内容如下：<br>
>{<br>
 "argv": [<br>
  "D:\\Anaconda\\envs\\tensorflow\\python.exe",<br>
  "-m",<br>
  "ipykernel_launcher",<br>
  "-f",<br>
  "{connection_file}"<br>
 ],<br>
 "display_name": "Python 3",<br>
 "language": "python"<br>
}<br>

　　可以看到里面指定了Python解释器的位置，修改成指定的Python环境路径保存，重启Jupyter Notebook即可。<br>
　　另外，如果再Jupyter Notebook中如果想要查看函数的具体格式的话，可以使用 Shift + Tab快捷键。<br>
#### 4、Jupyter 服务化<br>
　　有两种方法可以实现在Windows下创建一个JupyterNotebook的服务：<br>
　　一、利用pywin32这个python库，用python写一个服务安装脚本来实现jupyter服务（网上已有很多教程）；<br>
　　二、使用instsrv.exe与srvany.exe创建服务。<br>
##### 4.1 方式二说明<br>
1、安装instsrv.exe & srvany.exe(windows服务注册程序)<br>
　　复制instsrv.exe、srvany.exe文件到C:\Windows\SysWOW64文件夹下。<br>
　　命令行输入（cd 到SysWOW64文件夹下）：<br>
>instsrv.exe　Jupyter(服务名称)　C:\Windows\SysWOW64\srvany.exe　　　　# (必须是绝对路径)<br>

2、配置Jupyter notebook注册表<br>
　　(1) 打开注册表，在其中的HKEY_LOCAL_MACHINE\SYSTEM\CurrentControlSet\Services目录下找到刚刚创建的名为Jupyter的服务项，并在其下建立名为Parameters的项；<br>
　　(2) 在Parameters项中添加Application字符串与AppParameters字符串：<br>
　　　　Application字符串填写为jupyter-notebook的执行程序地址（一般为Python或Anaconda安装目录下的Scripts目录中：Ｄ:\Anaconda3\Scripts\Jupyter-notebook.exe）。<br>
　　　　AppParameters字符串中填写执行程序的参数（在这里指定jupyter的配置文件地址:--config="C:\Users\Lele\.jupyter\jupyter_notebook_config.py" --no-browser）。<br>
3、映射jupyter notebook服务端口<br>
> ssh -L 8090:127.0.0.1:8090    (修改了jupyter notebook的服务端口，因为与抓包工具冲突)

4、注册服务登陆权限<br>
　　最好修改服务的登录身份为“网络服务”或者“本地服务”，避免使用默认的“本地系统”账户（导致新建jupyter文件Kernel error:'帐户名与安全标识间无任何映射完成。'）运行JupyterNotebook。<br>

#### 6、Jupyter配置
##### 6.1、设置jupyter默认启动文件夹<br>
　　Windows下jupyter notebook默认的启动路径就是当前cmd启动jupyter 的路径：<br>
>C:\Users\用户名>jupyter notebook<br>

　　此时jupyter 的启动工作路径就是在　c:\Users\用户名，当更改盘符或者文件夹路径时　jupyter 的启动也会随着改变，为了使每次都在一个指定的工作路径下打开，可按如下设置：<br>
　　1) 打开命令行（win+R，输入cmd,然后按回车键),输入：<br>
>jupyter notebook --generate-config

　　此时会生成一个 jupyter_notebook_config.py 的文件，文件路径在 C:\Users\用户名\ .jupyter\ 下。<br>
　　2) 打开 jupyter_notebook_config.py 文件，找到
> \#\# The directory to use for notebooks and kernels.<br>
\#c.NotebookApp.notebook_dir = ''<br>

　　将要指定的路径填入 #c.NotebookApp.notebook_dir = '   ' 中的 '   ' 中即可，同时注意将 #c 中的 # 删去，：<br>
>eg: c.NotebookApp.notebook_dir = 'D:\jupyterProject'

　　3) 保存，在命令行中再次输入 jupyter notebook 指令即可在指定的文件夹下启动<br>

### 四、魔法方法
使用魔法函数可以简单的实现一些单纯python要很麻烦才能实现的功能。<br>
>%：行魔法函数，只对本行代码生效。<br> 
%%：Cell魔法函数，在整个Cell中生效，必须放于Cell首行。<br> 
%lsmagic：列出所有的魔法函数 <br>
%magic查看各个魔法函数的说明 <br>
?后面加上魔法函数名称，可以查看该函数的说明；如：?random查看random库文档<br>
在一个库、方法或变量前加上 ?，就可以获得它的一个快速语法说明<br>
使用分号可以阻止该行函数的结果输出<br>

一些常用魔法函数的示例：<br>

|魔法函数|作用|
|:--- |:---|
|%%writefile|调用外部python脚本|
|%run|调用外部python脚本|
|%timeit|测试单行语句的执行时间|
|%%timeit|测试整个单元中代码的执行时间|
|% matplotlib inline|显示 matplotlib 包生成的图形|
