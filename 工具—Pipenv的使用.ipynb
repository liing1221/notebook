# 一、pipenv简介
## 1.1、虚拟环境
　　虚拟环境是用于依赖项管理和项目隔离的python工具，它可以将python程序和pip包管理工具安装在本地的隔离目录中（非全局安装）。  
　　在实际开发中，不同项目，可能需要的python版本或第三方依赖包的版本不同，因此需要使用到虚拟环境来管理不同的项目环境。  
## 1.2、pipenv简介
　　Pipenv 是 Python.org 官方推进的 Python **包管理工具**，旨在将所有包管理工具（如 bundler、composer、npm、cargo、yarn 等）的优点集中应用于 Python 领域中的工具。它对各个平台都有很好的支持。  
　　Pipenv 可以为你的项目自动创建和管理虚拟环境，并且在你安装/删除包时自动在 Pipfile 中添加/移除相应的包。它同时生成一个非常重要的 Pipfile.lock 文件，来保证所构建环境的确定性。  
## 1.3、pipenv优点
- 无需再单独使用 pip 和 virtualenv ，在 Pipenv 中他们合并了。  
- 使用 requirements.txt 文件可能会产生很多问题，所以 Pipenv 使用 Pipfile 和 Pipfile.lock 将其代替，而且 Pipfile 如果不存在的话会自动创建，而且在安装、升级、移除依赖包的时候会自动更新 Pipfile 和 Pipfile.lock 文件。  
- 广泛使用 Hash 校验，保证安全性。  
- 通过图形更加清晰的深入了解当前的依赖关系(命令：pipenv graph) 。 
- 通过加载 .venv 文件简化开发流程 。 
## 1.4、pyenv、pipenv、virtualenv区别
- Pyenv: Python版本控制管理工具，可在系统中安装多个python版本，并且不影响系统自带的python环境，同时通过pyenv我们可以快速的在多个python版本之间来回切换。pyenv的原理就是对python版本进行统一管理，之后通过修改环境变量来切换python命令的指向。  
- Pipenv: Python 包管理工具，其创建虚拟环境可用于隔离不同项目的第三方库依赖，是对virtualenv的高级封装。  
- virtualenv: Python 自带的环境管理工具，比较低级。

# 三、pipenv工作流
## 3.1、准备工作
> cd myproject  
mkdir .venv    
\# 该步骤会使项环境依赖安装到 .venv 目录下；否则会安装到用户主目录的 .virtualvenv 目录内 

## 3.2、创建虚拟环境
>pipenv --python 3.9.7  # 或 pipenv --three  

　　使用指定的python版本(需本地已安装该版本python环境)初始化虚拟环境    
　　这会在项目目录中创建两个新文件：    
+ Pipfile:该文件是TOML格式，存放当前虚拟环境的配置信息，包括python版本，pypi源以及依赖包等，pipenv根据该文件寻找项目的根目录。  
+ Pipfile.lock:该文件是对Pipfile的锁定，支持锁定项目不同版本所依赖的环境。  

## 3.3、创建添加项目依赖环境
>pipenv install　　　# 如果存在Pipfile，则从中安装所有依赖项，并自动创建虚拟环境。   
pipenv install requests　　# 向项目环境中添加/安装一个依赖库(默认情况下会加锁，速度很慢)  
pipenv install \<pkg\> --skip-lock　　　# 安装第三方依赖但不加锁（加快速度）  
pipenv lock　　　　# 生成pipfile.lock(pipenv install 默认会生成pipfile.lock文件)  


　　上述命令用来安装项目特定的环境依赖，同时会生成Pipfile和Pipfile.lock文件。这样安装的依赖只对使用该虚拟环境的项目生效。  
　　使用pipenv创建虚拟环境，自动生成一个随机的虚拟环境目录名。  
　　如果在windows系统下执行命令，生成的虚拟环境在C:\Users\用户名\.virtualenvs文件夹下。  
　　一般虚拟环境目录名的前缀是你创建环境时所在的项目目录名，如在myblog目录下执行命令，虚拟环境的目录名称就是myproject-Gtn4e1q9，后半部分为随机字符串。  
　　pipenv使用 Pipfile 和 Pipfile.lock 来管理依赖包，并且在使用pipenv添加或删除包时，自动维护 Pipfile 文件，同时生成 Pipfile.lock 来锁定安装包的版本和依赖信息。相比pip需要手动维护requirements.txt 中的安装包和版本，具有很大的进步。这两个文件的区别是 Pipfile中安装的包不包含包的具体版本号，而Pipfile.lock是包含包的具体的版本号的。如果不想产生Pipfile.lock文件，在安装依赖包的时候，加上–skip-lock选项即可。  
　　打开依赖包文件Pipefile，可以看到python_version = “3.7”，说明这个项目环境是基于Python 3.7版本的。package部分列出来了项目的依赖包，　如：pytest = “\*”，星号代表最新版本。  
　　Pipfile.lock是包含安装的依赖包具体的版本号，可以看到本次安装的pytest是5.4.1版本，并且它的依赖包的版本也列出来了。  

## 3.4、提高依赖安装速度
　　在使用pipenv的时候，常常会安装过程比较慢，通过加上 -v 参数，可以看到安装过程中的步骤信息，卡在了下载那里，这时应该可以意识到是因为网络的原因，pipenv创建的 Pipfile 中默认的Pypi源是python官方源，我们国内用户访问下载的时候会很慢。  
　　如果想通过Pypi的国内镜像安装依赖包，可以在安装软件包时，指定：**–pypi-mirror**，比如通过清华大学镜像安装flask软件包：  
>pipenv install --pypi-mirror https://pypi.tuna.tsinghua.edu.cn/simple flask

　　为了避免每次都要指定–pypi-mirror，我一般会在创建好Pipfile以后，将文件中 source 块下的 url 字段，设置为国内的 pypi 源，我推荐的是清华的Pypi源或者阿里源，具体设置如下：  
```
[[source]]  
url = "https://pypi.tuna.tsinghua.edu.cn/simple"  
verify_ssl = true  
name = "pypi"  
```
## 3.5、激活虚拟环境
>pipenv shell　　　# activate  
pipenv --version  
exit　　　# deactivate     

　　激活环境后会自动进入到虚拟环境中(通过生成一个新的shell子进程)。  
　　退出虚拟环境时，执行exit 或 ctrl + d 即可（推出该shell子进程）。  

## 3.6、删除依赖包或环境
　　如果是要删除虚拟环境中的第三方包，执行  
>pipenv uninstall pytest  
pipenv uninstall --all　　　　# 删除所有包  
pipenv --rm　　　　　　　# 删除虚拟环境  

## 3.7、安装指定版本依赖包
　　根据Pipfile安装，是安装其列出来的第三方包的最新版本。如果是想安装Pipfile.lock中指定的固定版本的第三方依赖包，需要执行：    
>pipenv install --ignore-pipfile    
pipenv install　　　# 如果存在Pipfile，则从中安装所有依赖项，并自动创建虚拟环境    
pipenv install --dev　　　# 安装开发环境依赖(如py.tests,mock等)  
pipenv install [package_name]　　　# 指定包名  
pipenv install -r requirements.txt　　　# 如果项目已经存在requirements.txt  


　　pipenv install有多重作用：  
+ 如果虚拟环境已经存在，则安装Pipfile中的依赖包;  
+ 如果虚拟环境不存在，但Pipfile存在，则根据Pipfile中python版本创建虚拟环境并安装依赖包;  
+ 如果虚拟环境和Pipfile都不存在，则根据系统默认python版本创建虚拟环境.  

　　另外你也可以以下面格式的URL安装在git或其他版本控制系统中的包。
```shell
# <vcs_type>+<scheme>://<location>/<user_or_organization>/<repository>@<branch_or_tag>#<package_name>
```
　　**vcs_type有效值：**　git，bzr，svn，hg  
　　**scheme有效值：**　http，https，ssh，file  
　　**branch_or_tag：**　可选参数  

　　强烈建议以编辑模式安装任何版本控制依赖，如下示例：  

>pipenv install -e git+https://github.com/requests/requests.git@v2.19#egg=requests　　# 安装requests




## 3.8、直接使用虚拟环境运行命令
>pipenv run python xxx.py  

## 3.9、Pipenv 依赖更新
>\# 查看所有需要更新的包。  
pipenv update --outdated   
\# 更新包  
pipenv update　　　　　　　　　　# 更新所有的依赖项  
pipenv update \<pkg\>　　　　　　# 升级指定的包。

## 3.10、查看依赖信息
```shell
pipenv --venv　　　　# 查看虚拟环境路径地址  
pipenv --where　　　　# 查看本地项目路径  
pipenv --py　　　　# 查看python解释器路径  
pipenv graph　　　　# 以树结构查看所有依赖包信息（包括子依赖包）或说依赖关系查看  
pipenv open <phg>　　　# 查看具体依赖包的信息  
pipenv check　　　　　# 查看依赖包的安全性  
pipenv check --style xxx.py　　　# pipenv默认集成了flake8, 可以用来检测编码风格  
```
## 3.11、pipenv与requirements.txt
### 3.11.1、导出环境到requirements.txt
>pipenv lock -r > requirements.txt  
pipenv lock -r --dev > requirements.txt　　　# 只导出开发(dev-packages)用的包(pipenv lock -r -d)  
pipenv run pip freeze > requirements.txt

### 3.11.2、从requirements.txt导入环境
　　如果在运行 pipenv install 命令时有一个可用的 requirements.txt 文件，pipenv会自动将这个文件中的内容导出并创建一个 Pipfile 文件。  
　　如果 requirements.txt 文件中的包有固定的版本号，你可以通过编辑新生成的 Pipfile 文件来移除那些版本号，并让 pipenv 来负责处理。如果你想在 Pipfile 中保留那些固定的版本号，可以使用 pipenv lock --keep-outdated 命令。  
>pipenv install -r requirements.txt

## 3.12、建议&版本控制
- 请在版本控制中保留 Pipfile 和 Pipfile.lock 文件  
- 如果目标是多个Python版本，则不要将 Pipfile 保存在版本控制系统中  
- 在 Pipfile 的 [require] 部分中指定目标Python版本。理想情况下，您应该只有一个目标Python版本，因为这是一个部署工具。  
- pipenv install 命令完全兼容 pip install 语法。  
- Pipfile 使用了 TOML 规范。  


# 四、线上部署
## 4.1、线上使用virtualenv管理虚拟环境
　　可以直接在virtualenv的虚拟环境中安装pipenv，这样就可以直接运行我们基于pipenv构建的项目了
```sh
# 创建虚拟环境  
python3 -m virtualenv .venv  
# 用虚拟环境的pip安装pipenv  
.venv/bin/python -m pip install -U pip pipenv  
# 通过虚拟环境的pipenv安装第三方包  
.venv/bin/pipenv install --dev --deploy  
```
## 4.2、线上无virtualenv工具
　　通过pipenv导出requirements.txt，然后到线上安装第三方包  
```sh
# 本地执行，然后导到线上  
pipenv run pip freeze > requirements.txt  

# 服务器安装第三方包  
python -m pip -r requirements.txt  
```
