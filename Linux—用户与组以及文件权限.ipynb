## Linux—用户与组以及文件权限

### 1、用户、组及文件权限常用命令：
>用户	　　who, whoami, su, useradd, userdel, passwd, usermod, /etc/passwd<br>
组	　　　groupadd, groupdel, groupmod, /etc/group<br>
文件	　　chmod, chown, chgrp<br>
其他	　　sudo, exit<br>

### 2、查看已登录用户
　　**显示当前所有已登陆到系统的用户信息。**<br>
> who<br>

　　参数:
>-H    显示各列的标题（表头）。<br>
-u    显示闲置时间。如果用户在前1分钟内有操作, 显示“.”; 超过24小时没有操作, 显示“old”字符串。<br>
-m    显示当前会话的用户的信息, 相当于: who am i<br>
-q    显示当前所有已登录的 账户名称 和 总人数。<br>
-T    显示用户的 状态信息<br>

　　例子:
>who         // 显示当前所有已登录的用户<br>
who -q      // 显示当前所有已登录的账号名称和总人数<br>
who am i    // 显示当前会话的用户的信息<br>
who -m      // 显示当前会话的用户的信息<br>
who -uTH    // 显示当前所有已登录的用户信息（包括 闲置时间, 用户状态, 各列标题）<br>

　　**显示当前会话的用户的名称:** <br>
>whoami<br>

　　**切换用户：**<br>
>su<br>

### 3、创建用户账户与密码：
前提，使用管理员账户登录，或获取管理员权限。<br>
**useradd命令：**<br>
格式: <br>
>useradd [-potions] username

参数:<br>
>-d    指定新建用户的主目录, 如果不指定, 则系统自动在 /home 目录创建一个和用户名相同的文件夹作为新建用户的主目录（/home/username）<br>
-m    自动创建主目录文件夹；如果存在不再创建，但是此目录并不属于新创建用户；如果主目录不存在，则强制创建； -m和-d一块使用。<br>
-s：  指定用户登录时的shell版本<br>
-M：  不创建主目录<br>
-g    指定新建用户所在组名称, 如果不指定, 则系统自动创建一个和用户名相同的组作为新建用户所属组<br>

注意:    新创建的用户默认不能使用 sudo 命令, 需要另外添加该命令的权限, 实际上 sudo 是一个组的名称, 把组sudo 追加到用户的附加组列表, 该用户便能在输入用户自己密码的情况下执行sudo 命令。<br>
　　例如让用户 user01 拥有 sudo 命令权限, 命令格式为: usermod -aG sudo user01<br>
例如：<br>
sudo  useradd  -d  "/home/tt"   -m   -s "/bin/bash"   tt<br>
解释：   -d   “/home/tt" ：就是指定/home/tt为主目录  -m   就是如果/home/tt不存在就强制创建 -s    就是指定shell版本<br>

useradd：需要使用参数选项指定上述基本设置，如果不使用任何参数，则创建的用户无密码、无主目录、没有指定shell版本。<br>
1）创建新用户：useradd + 用户名<br>
2）为新用户指定用户密码：passwd 123456   设置密码只需按照提示操作即可，密码如果过于简单的话，会提示简单，不用管， 直接再输入一次就会更新成功。<br>
3）需要为用户指定shell版本和用户登录主目录。<br>
补充：查看shell版本 echo shell<br>
查看shell版本，发现是/bin/sh：说明这是没有指定shell版本<br>

**adduser命令：**<br>
adduser：会自动为创建的用户指定主目录、系统shell版本，会在创建时输入用户密码。此方法不用添加-m参数，同样是可以在home目录下生成该用户的家目录，并在创建此用户的时候，并为该用户设置了密码不需要在另行设置。<br>
>adduser + 用户名　　　　# 这样在创建用户名时，就创建了用户的主目录以及密码。<br>

默认情况下：<br>
adduser在创建用户时会主动调用：/etc/adduser.conf；<br>
在创建用户主目录时默认在/home下，而且创建为 /home/用户名。<br>
如果主目录已经存在，就不再创建，但是此主目录虽然作为新用户的主目录，而且默认登录时会进入这个目录下，但是这个目录并不是属于新用户，当使用userdel删除新用户时，并不会删除这个主目录，因为这个主目录在创建前已经存在且并不属于这个用户。<br>
为用户指定shell版本为：/bin/bash<br>
因此常用参数选项为：<br>
>--home：                       指定创建主目录的路径，默认是在/home目录下创建用户名同名的目录，这里可以指定；如果主目录同名目录存在，则不再创建，仅在登录时进入主目录。<br>
--quiet：                        即只打印警告和错误信息，忽略其他信息。<br>
--debug：                     定位错误信息。<br>
--conf：                         在创建用户时使用指定的configuration文件。<br>
--force-badname：     默认在创建用户时会进行/etc/adduser.conf中的正则表达式检查用户名是否合法，如果想使用弱检查，则使用这个选项，如果不想检查，可以将/etc/adduser.conf中相关选项屏蔽。

### 4、更改用户密码:
**passwd命令**<br>
设置或修改用户密码, 普通用户可以根据原密码修改自己的密码, 超级用户可以直接重置所有用户（包括 root 用户）的密码。<br>
格式:<br>
>passwd [-options] [username]

参数:<br>
>-a    all, 此选项只能和 -S 一起使用, 来显示所有用户的状态<br>
-d    delete, 删除用户密码（把密码置为空）<br>
-l    lock, 锁定指定账户（将密码更改为一个不可能与加密值匹配的值来禁用）<br>
-u    unlock, 解锁指定账户<br>
-S    status, 显示账户状态信息<br>

例子:<br>
>passwd              // 更改自己的密码, 输入 原密码, 新密码, 确认新密码<br>
passwd user01       // 重置用户的密码, 输入 新密码, 确认新密码<br>
passwd -aS          // 查看所有用户的状态<br>
passwd -S user01    // 查看用户 user01 的状态<br>

### 5、修改用户信息: 
**usermod命令**<br>
usermode 命令用于修改用户的基本信息。不允许修改正在线上的用户的名称, 不允许修改正在系统上执行程序的用户的ID。<br>
格式: <br>
>usermod [-options] username

参数:<br>
>    -c    comment, 设置帐户注释<br>
    -d    home_dir, 设置用户主目录<br>
    -e    expiredate, 设置帐户过期日期<br>
    -g    group, 强制修改用户的新主组<br>
    -G    groups, 修改用户新的附加组列表<br>
    -a    append group, 将用户追加至上边 -G 中提到的附加组中, 并不从其它组中删除此用户<br>
    -u    uid, 修改账户的用户ID<br>
    -l    login, 设置用户的登录名称<br>
    -s    shell, 修改用户登录后使用的 shell<br>
    -L    lock, 锁定用户帐户（等同于 passwd -l 命令锁定账户）<br>
    -U    unlock, 解锁用户帐户（等同于 passwd -u 命令解账户）<br>

例子:<br>
>   usermod -g group01 user01   // 把用户 user01 的主组修改为 group01 组<br>
    usermod -u 9999 user01      // 把用户 user01 的 ID 修改为 9999<br>
    usermod -l u1 user01        // 把用户 user01 的名称修改为 u1<br>
    usermod -aG sudo user01     // 把 组sudo 添加到用户 user01 的附加组列表, 使该用户拥有执行 sudo 命令的权限<br>

**用户账户信息: /etc/passwd**<br>
文件 /etc/passwd 中存放了所有用户的账户信息, 每一行为一个用户, 每个用户的账户信息分为 7 个字段, 用 “:” 分隔。<br>
格式:<br>
    Username : Password : User ID : Group ID : Comment : Home Diretory : Login Command<br>
    用户名　　: 密码　　　: 用户ID　: 组ID　　: 用户注解　 : 主目录　　　: 登录后执行的命令<br>
其中 密码字段 只显示一个特殊字符“x”或“\*”, 加密后的密码存放在 /etc/shadow 文件中, 只有超级用户才能访问。<br>

### 6、组管理
**添加组: groupadd**<br>
格式: <br>
>groupadd [-options] groupname

参数:<br>
>-g    指定新建组的ID<br>

例子:<br>
>    groupadd group01            // 新建名称为 group01 的组
    groupadd -g 1001 group02    // 新建名称为 group02 的组,  并给指定组的 ID 为 1001

**删除组: groupdel**<br>
删除组, 如果组内还包含有用户, 需要想将用户删除或移出该组后才能删除。<br>

格式:<br>
>groupdel groupname

**改变组信息: groupmod**<br>
格式:<br>
>groupmod [-options] groupname

参数:<br>
>    -n    修改组的名称

例子:<br>
>    groupmod -n NewGroupName OldGroupName

**查看所有组: /etc/group**<br>
文件 /etc/group 中存放了所有组的信息, 每一行表示一个组, 每个组的信息分为 4 个字段, 用 “:” 分隔。<br>
格式:<br>
>    Group Name : Password : Group ID : User List

**用户组用法**<br>
1.创建组：sudo groupadd  组名<br>
2.查看组：cat  /etc/group<br>
3.删除用户组： sudo groupdel 组名<br>
4.修改用户所在组（创建组名，创建用户）<br>
例如创建了两个组 xxx 和yyy ，创建了两个用户xiaohua  xiaocao<br>
① sudo  usermod  -G   xxx    xiaohua       将用户xiaohua放进xxx组内<br>
    sudo   usermod  -G   yyy  xiaocao<br>
② cat  /etc/group   |  grep xxx       查询xxx组<br>
或   cat  /etc/group   |  grep xiaohua （用户名）     两种方法都行<br>
③ 查询用户在哪些组：groups   xiaohua（用户名）<br>
④ xiaohua 本来在xxx组内，也可以添加到yyy组内，相当于（一个用户名可以在多个组中，也就是说一个组可存多个用户名）<br>
sudo  usermod  -a  -G  yyy  xiaohua         将xiaohua用户名添加到yyy组内<br>
groups  xiaohua                                         查看即可<br>

### 7、userdel删除用户命令:<br>
删除用户时, 如果用户所属组是创建用户时自动创建的和用户名称同名的组, 并且该组内没有其他用户, 则该组也会被删掉。<br>
格式:<br>
>userdel [-options] username<br>

参数:<br>
>-f    强制删除用户, 即使用户当前已登录
-r    删除用户, 同时删除用户主目录

1）只删除用户：<br>
>sudo   userdel   用户名     # 删除用户但不会删除主目录，文件还存在，但相关的用户信息没有了<br>

2）连同用户主目录一块删除：<br>
>sudo  userdel   -r   用户名    # 删除用户，同时删除用户主目录，文件夹也被删掉了（还是用这个多）<br>

如果创建时主目录已经存在，即主目录不属于当前要删除的用户，则无法删除主目录。<br>
如果用了第一种方法，没有加-r删除的用户，那么主目录不易被删掉，解决方法：再重新添加一遍这个用户，用户名一致，再删掉就ok；其次如果删不掉用户名，就关掉终端窗口再重新删除即可。<br>
**用户相关文件：**<br>
>/etc/passwd - 使用者帐号资讯，可以查看用户信息<br>
/etc/shadow - 使用者帐号资讯加密<br>
/etc/group - 群组资讯<br>
/etc/default/useradd - 定义资讯<br>
/etc/login.defs - 系统广义设定<br>
/etc/skel - 内含定义档的目录<br>

### 8、su 命令切换用户
1)、su  目录名（su   zz格式）　　　　# 用户会切换但还在原用户路径下，此时切换后可cd 指令回到家目录即可<br>
2)、 su - 目录名       # 会直接切换到用户的家目录，密码要输入要切换的用户密码<br>
总结：su和su– 这两个命令区别在于：su - 切换到对应的用户时会将当前的工作目录自动转换到切换后的用户主目录。<br>
3)、切换到超级管理员的方法，只记一种即可：（前面都加sudo）只记sudo su -  ，切换到普通用户 su - 用户名<br>
>su/su root　　　　切换到root用户
su -/su - root　　　切换到root用户，同时切换目录到/root
su 普通用户/su - 普通用户　　　　　　切换到普通用户，同时到普通用户所在目录

注意：普通用户（命令添加的用户）由于没有sudo权限所以暂时切换不到root账户。<br>
root > 创建虚拟机时的用户（zxk）>自己添加的用户<br>
要提升root权限：先返回到高一级例在zxk用户名下执行下面命令：<br>
sudo usermod -a -G sudo xiaocao（自己添加的用户名）<br>
sudo usermod -a -G adm xiaocao<br>
接着转换到xiaocao用户名上：指令：su - xiaocao<br>
再输入指令： sudo su -    可转换到root权限了。<br>

### 10、其他
以其他身份来执行命令: sudo
sodu, 以其他身份来执行命令, 默认预设的身份为 root 用户。普通用户身份可能没有权限执行某些命令或查看某些文件/文件夹, 此时需要切换到 root 用户, 如此比较麻烦, 也会暴露 root 用户密码给更多的普通用户, 因此可以为用户添加 sudo 权限（详见上面的 usermod 命令介绍）, 让该用户可以临时以 root 身份来执行相关命令。<br>
执行该命令需要输入当前登录用户的密码。<br>
格式: <br>
>sudo [-options] original_commands

参数:<br>
>    -H　　　　将HOME环境变量设为新身份的HOME环境变量。<br>
    -u　　　　以指定用户的身份来执行命令, 默认为 root 用户。<br>

例子:<br>
>sudo ls /root       // 以 root 身份查看 /root 目录
