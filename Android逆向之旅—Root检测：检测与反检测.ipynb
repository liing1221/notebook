首先 我们更新到最新稳定版magisk(面具)24.3


点击右上角设置，找到Zygisk选项并且打开，打开后重启手机


重启手机后刷入Shamiko模块，重启手机，打开面具右上角设置，打开遵循排除列表选项，再点击配置排除列表


打开排除配置列表后，勾选你要隐藏root的应用，然后关闭排除列表选项


如下图所示，如果你不关比排除列表选项，那么Shamiko模块就会不工作，隐藏root就会不生效，所以是必须关闭的，首次配置建议重启一次手机，已达到最佳的隐藏效果。



一般来说操作完上面的步骤就可以了，如果发现有app还是会提示设备已root，那么就可以继续下面的步骤。

此时我们要随机包名Magisk（面具）达到隐藏Magisk（面具）的效果，以防止被某些软件检测到。


如果遇到下图，老是转圈的情况，建议使用魔法（科学上网）进行操作，至于怎么找魔法，我就不多说，因为不可抗力因素，不能放出来，就看你们找了。


隐藏Magisk（面具）后，此时大部分的软件都已经检测不到root，但是还有一部分顽强的软件还是可以检测到root十分顽强，那么我们该怎么办呢，下面的硬货又来了。

我们下载Zygisk版本的LSPosed，用面具刷入，重启手机



下载隐藏应用列表


然后打开隐藏应用列表，点击检测测试下载测试软件，后面需要用到和点击安装magisk插件，选择Zygisk，安装成功后，重启手机


然后点击模板管理，创建黑名单模板，填写以下可能被检测到的maps规则


如下图将规则复制粘贴：

magisk

root

lsposed

edxposed

xposed

qn_dyn_lib

qn_mmkv

org.meowcatedxposedmanager

org.lsposed.manage

manager

com.topjohnwumagisk

magisk manager


然后然后点击隐藏应用，勾选带有root权限的所有软件


保存刚才设置好的规则


选择你要隐藏root的软件，按照下面的图片并且保存选择刚才设置好的规则模板，保存即可。


还要注意的是，那个工作模式不要盲目选择所有隐藏方式，容易让某些软件直接崩溃 下面打开测试软件，我们会发现，没有被检测到root，这样说明很多软件都不会被检测了，可以正常使用。

以上就是新版magisk24.0+隐藏面具的方法了，magisk的延伸版本 Canary（金丝雀版本）和alpha（阿尔法版本）隐藏方法同理，还要注意一点Zygisk不能和riru系列的模块共存，安装riru，Shamiko模块就会不工作，如果你想用riru系列模块又想隐藏root建议会退magisk23.0即可。
