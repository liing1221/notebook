## Jupyter notebook—使用之公式插入
　　参考：[Jupyter-Notebook插入公式](https://blog.csdn.net/weixin_41923961/article/details/82761639)<br>
　　　　　[Markdown公式编辑学习笔记](https://www.cnblogs.com/q735613050/p/7253073.html#jump)<br>
jupyter notebook是一个python的交互式开发环境，广泛应用于数据分析的场景下。<br>
在jupyter notebook中，还可以很方便的编辑数学公式。<br>
### 1、Markdown状态:<br>
编辑公式，首先要使单元格成为Markdown状态。

### 2、公式的行内（inline）模式与单独显示（display）模式：
公式的行内模式为：\\$ 数学公式 \\$ ，公式将与行内其前后的文字混排。<br>
>一元一次方程：\\$y = a + bx\\$<br>
显示：$y = a + bx$

公式的单独显示模式为：\\$\\$数学公式\\$\\$<br>
>\\$\\$y = ax + b\\$\\$<br>
显示：$$y = ax + b$$

### 3、输入上下标<br>
^表示上标, _ 表示下标。如果上下标的内容多于一个字符，需要用 {}将这些内容括成一个整体。上下标可以嵌套，也可以同时使用。<br>
>\\$y=ax{^2_{ij}}+b\\$<br>
显示：$y=ax{^2_{ij}}+b$

另外，如果要在左右两边都有上下标，可以用\sideset 命令：
>\\$\\$ \sideset{^1_2}{^3_4}\bigotimes \\$\\$<br>
显示：$$ \sideset{^1_2}{^3_4}\bigotimes $$<br>

### 4、输入括号与分隔符<br>
()、[] 和 | 表示符号本身，使用 \\{\\} 来表示 {}。当要显示大号的括号或分隔符时，要用 \left 和 \right 命令。<br>
一些特殊的括号：<br>

|输入|显示|
|---|:---:|
|\\$\\$\langle 表达式 \rangle\\$\\$|$$\langle 表达式 \rangle$$|	
|\\$\\$\lceil 表达式 \rceil\\$\\$|$$\lceil 表达式 \rceil$$|
|\\$\\$\lfloor 表达式 \rfloor\\$\\$|$$\lfloor 表达式 \rfloor$$|
|\\$\\$\lbrace 表达式 \rbrace\\$\\$|$$\lbrace 表达式 \rbrace$$|
>\\$\\$ f(x,y,z) = 3y^2z \left( 3+\frac{7x+5}{1+y^2} \right) \\$\\$<br>
显示：$$ f(x,y,z) = 3y^2z \left( 3+\frac{7x+5}{1+y^2} \right) $$

有时候要用\left.或\right.进行匹配而不显示本身。<br>
>\\$\\$\left. \frac{ {\rm d}u}{ {\rm d}x} \right| _{x=0}\\$\\$<br>
显示：$$\left. \frac{ {\rm d}u}{ {\rm d}x} \right| _{x=0}$$

### 5、输入分数
通常使用 \frac {分子}{分母} 命令产生一个分数 \frac {分子}{分母}，分数可嵌套。<br>
便捷情况可直接输入 \frac ab 来快速生成一个\frac ab。<br>
如果分式很复杂，亦可使用 分子\over 分母 命令，此时分数仅有一层。<br>
>\\$\\$\frac{a-1}{b-1} \quad and \quad {a+1\over b+1}\\$\\$<br>
显示：$$\frac{a-1}{b-1} \quad and \quad {a+1\over b+1}$$

### 6、输入开方
使用 \sqrt [根指数，省略时为2]{被开方数} 命令输入开方。<br>
>\\$\\$\sqrt{2} \quad and \quad \sqrt[n]{3}\\$\\$<br>
显示：$$\sqrt{2} \quad and \quad \sqrt[n]{3}$$

### 7、输入省略号
数学公式中常见的省略号有两种：\ldots 表示与文本底线对齐的省略号；\cdots 表示与文本中线对齐的省略号。<br>
>\\$\\$f(x_1,x_2,\underbrace{\ldots}_{\rm ldots} ,x_n) = x_1^2 + x_2^2 + \underbrace{\cdots}_{\rm cdots} + x_n^2\\$\\$<br>
显示：$$f(x_1,x_2,\underbrace{\ldots}_{\rm ldots} ,x_n) = x_1^2 + x_2^2 + \underbrace{\cdots}_{\rm cdots} + x_n^2$$

### 8、输入矢量<br>
使用 \vec{矢量}来自动产生一个矢量。也可以使用 \overrightarrow等命令自定义字母上方的符号。<br>
>\\$\\$\vec{a} \cdot \vec{b}=0\\$\\$<br>
显示：$$\vec{a} \cdot \vec{b}=0$$
\\$\\$\overleftarrow{xy} \quad and \quad \overleftrightarrow{xy} \quad and \quad \overrightarrow{xy}\\$\\$<br>
显示：$$\overleftarrow{xy} \quad and \quad \overleftrightarrow{xy} \quad and \quad \overrightarrow{xy}$$

### 9、输入积分
使用 \int_积分下限^积分上限 {被积表达式} 来输入一个积分。<br>
>\\$\\$\int_0^1 {x^2} \,{\rm d}x\\$\\$<br>
显示：$$\int_0^1 {x^2} \,{\rm d}x$$

### 10、输入极限运算
使用\lim_{变量 \to 表达式} 表达式 来输入一个极限。如有需求，可以更改 \to 符号至任意符号。<br>
>\\$\\$ \lim_{n \to +\infty} \frac{1}{n(n+1)} \quad and \quad \lim_{x\leftarrow{示例}} \frac{1}{n(n+1)} \\$\\$<br>
显示：$$ \lim_{n \to +\infty} \frac{1}{n(n+1)} \quad and \quad \lim_{x\leftarrow{示例}} \frac{1}{n(n+1)} $$

### 11、输入累加、累乘运算
使用 \sum_{下标表达式}^{上标表达式} {累加表达式}来输入一个累加。<br>
与之类似，使用 \prod \bigcup \bigcap来分别输入累乘、并集和交集。<br>
此类符号在行内显示时上下标表达式将会移至右上角和右下角。<br>
>\\$\\$\sum_{i=1}^n \frac{1}{i^2} \quad and \quad \prod_{i=1}^n \frac{1}{i^2} \quad and \quad \bigcup_{i=1}^{2} R\\$\\$<br>
显示：$$\sum_{i=1}^n \frac{1}{i^2} \quad and \quad \prod_{i=1}^n \frac{1}{i^2} \quad and \quad \bigcup_{i=1}^{2} R$$

### 12、大括号和行标的使用
使用 \left和 \right来创建自动匹配高度的 (圆括号)，[方括号] 和 {花括号}。<br>
在每个公式末尾前使用\tag{行标}来实现行标。<br>
>\\$\\$<br>
f\left(<br>
　　\left[ <br>
　　　\frac{<br>
　　　　1+\left\{x,y\right\}<br>
　　　}{<br>
　　　　\left(<br>
　　　　　\frac{x}{y}+\frac{y}{x}<br>
　　　　\right)<br>
　　　　\left(u+1\right)<br>
　　　}+a<br>
　　\right]^{3/2}<br>
\right)<br>
\tag{行标}<br>
\\$\\$<br>
显示：$$
f\left(
   \left[ 
     \frac{
       1+\left\{x,y\right\}
     }{
       \left(
          \frac{x}{y}+\frac{y}{x}
       \right)
       \left(u+1\right)
     }+a
   \right]^{3/2}
\right)
\tag{12}
$$

### 13、偏导
>\\$\\$\frac{\partial^{2}y}{\partial x^{2}}\\$\\$<br>
显示：$$\frac{\partial^{2}y}{\partial x^{2}}$$
