# Urllib库详解




### URL编码
　　urlencode 是针对字典(键值对)的操作，也可以单独对一个字符串值进行编码，基于是否把空格替换为加号 + 的不同行为，分别有  quote 和 quote_plus 操作。  
#### urlencode()
```python
urlencode(query, doseq=False, safe='', encoding=None, errors=None, quote_via=quote_plus)
```
　　参数： 
+ query：将要编码为URL查询字符串的字典或双元素元组序列。
+ doseq: doseq为true时，如果查询参数中的任何值是序列，则每个序列元素被转换为一个单独的参数。如果查询参数是一个双元素元组序列，则输出参数的顺序将与输入参数的顺序匹配。
+ quote_via：函数类型参数，接收参数为 safe、encoding、errors.(safe、errors仅用于查询参数的组件是str时)

　　注意：  
　　此处，特殊字符编码为 %xx 的形式，并且空格转换为加号 +，而不是编码为 %20, 这有可能造成解码时不一致而把加号 + 作为一个实际的字符来处理。如果希望空格编译编码为 %20, 而非加号 + 的话，调用  urlencode 函数可以由参数 quote_via 指定为  quote，它的默认值为 quote_plus，即空格替换为加号 +。

#### quote()
　　对字符串进行URL编码，空格编码为 %20。
```python
quote(string, safe='/', encoding=None, errors=None)
```
　　注意：
　　1. String和safe可以是STR或bytes对象。如果string是bytes对象，则不能指定Encoding和errors。  
　　2. 可选的encoding和errors参数指定如何处理str.encode方法所接受的非ascii字符。默认情况下，encoding='utf-8'(字符用utf-8编码)，errors='strict'(不支持的字符将引发UnicodeEncodeError)。

#### quote_plus()
　　对字符串进行URL编码，空格替换为加号 + , 原字符串中的加号将被转义，除非它们包含在safe中。
```python
quote_plus(string, safe='', encoding=None, errors=None)
```

```python
parse_qs(qs, keep_blank_values=False, strict_parsing=False,
             encoding='utf-8', errors='replace', max_num_fields=None, separator='&')
```

#### unquote()
　　与 quote 相对应的操作有 unquote ，也是单独对某个字符串的URL解码操作。
```python
unquote(string, encoding='utf-8', errors='replace')
```

#### unquote_plus()
　　与 quote_plus 相对应的操作有 unquote_plus ，也是单独对某个字符串的解码。
```python
unquote_plus(string, encoding='utf-8', errors='replace')
```

　　unquote 无法把加号 + 还原为空格，而使用 unquote_plus 好像总是更保险。是否有 URL 中的加号不进行编码呢，保持为 +？应该不会，加号 + 也不过是被编码为 %2B 而已。   
　　同时，unquote 或 unquote_plus 对 & 符号是无害的，所以它也能直接对整个查询字符串进行解码。
