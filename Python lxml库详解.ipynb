## Python lxml库详解
参考：[lxml官方教程](https://lxml.de/tutorial.html)<br>

### 一、lxml 安装介绍：
> pip install lxml

　　lxml is the most feature-rich and easy-to-use library for processing XML and HTML in the Python language.<br>
　　lxml 是提供了一个 Pythonic API ，并且她也是几乎完全兼容 ElementTree API 的。lxml 的开发团队也是力求减少开发者的学习成本，不过 lxml 的教程感觉一点都不友好。<br>
　　lxml 下面有 etree 、html 、cssselect 等众多模块。另外，<br>
　　lxml can make use of BeautifulSoup as a parser backend, just like BeautifulSoup can employ lxml as a parser.<br>
　　lxml interfaces with BeautifulSoup through the lxml.html.soupparser module.<br>
　　我们知道，BeautifulSoup 可以将 lxml 作为解析器，但是 lmxl.html.soupparser 也可以调用 BeautifulSoup。使用时，认准 BeautifulSoup 强大检测编码能力，并且能够以 utf-8 形式输出；lxml 具有速度飞起的文本解析力。<br>
　　lxml提供了以下几种功能：<br>
>**1、**romstring():解析字符串<br>
**2、**HTML():解析HTML对象<br>
**3、**XML():解析XML对象<br>
**4、**parse():解析文件类型对象<br>

### 二、lxml.etree：
#### 1、
　　lxml.etree, 作为一个强大的 xml 处理模块，它能将字符串或流读进来解析成 ElementTree ，又能将 ElementTree 转换为字符串或流。<br>
>etree.fromstring( string )<br>
　　　将 string 解析为 Element 或者 ElementTree 。<br>
etree.parse( file )<br>
　　　将文件或者是 file_like 对象解析为 ElementTree (not an Element object)，因为 parse() 一般解析整篇文档，字符串解析函数一般<br>
　　　只解析片段。其中 file 还可以是 HTTP/FTP URL ，也就是说，file 应该是一个 Bytes 流。不过 libxml2 中的 HTTP/FTP client 比较<br>
　　　简单，所以可以考虑使用第三方库。etree.parse()的第二个参数可用来指定解析器，如：etree.parse('test.html',etree.HTMLParser())，<br>
　　　指定解析器HTMLParser会根据文件修复HTML文件中缺失的如声明信息<br>
XML( text ) / HTML( text )<br>
　　　行为比较像 fromstring() ，比较直接的对 XML 和 HTML 文档进行特定解析，可以修改解析器 parser 参数，不同的解析器设置区别还是<br>
　　　蛮大的。其中 parser 可以由相应的 XMLParser() / HTMLParser() 函数生成，可设置项有很多不仅限于 encoding 、recover 、<br>
　　　remove_blank_text 、remove_comments ，但是默认值是存在不同的。比如 recover 在 XML 中是 False ，而在 HTML 中是 True 。在 <br>
　　　HTML 中会使用 libxml2 的 recover 算法对不合规的标签进行修复。<br>
etree.tostring(elt, encoding='utf-8'))<br>
　　　将一个 Element 或者 ElementTree 转换为 string 形式。这里面有几个可选参数：pretty_print=False 表示是否格式化提高可读性；<br>
　　　method="xml" 选择输出后的文档格式，不同的选择，做的修改也不相同，可选参数有 xml 、html 、text(文本内容不包括标签，就<br>
　　　是纯文本内容，tail也是) 、c14n (规范化 xml )；encoding=None 表示以什么编码的字符串输出，在无 xml 文档声明情况下默认是<br>
　　　ASCⅡ(汉字会以ACCII码显示) ，可通过 encoding=None 进行修改，但是如果所改编码不是 utf-8 兼容，那么将会启用默认声明。<br>
ElementTree.write( file )<br>
　　　这个是 ElementTree 特有的方法，是将 ElementTree 写到 a file, a file-like object, or a URL (via FTP PUT or HTTP POST) 。<br>
　　　可选参数和 tostring() 差不多，也有不同。<br>

　　剩下就是操作了，增删查改一样也不少，而查找是剩下三个的基础。<br>

删clear() remove()<br>
查find*() get*() iter*() xpath()<br>
改set() 以及属性 .tag .attrib .text .tail等的操作<br>

ElementTree 实例拥有的方法较少，常用的大概有 7 种，主要是查找find*() get*() xpath()。<br>

#### 2、Element ：
　　Element是XML处理的核心类，Element对象可以直观的理解为XML的节点，大部分XML节点的处理都是围绕该类进行的。这部分包括三个内容：节点的操作、节点属性的操作、节点内文本的操作。<br>
##### 1)、创建Element对象
   使用Element方法，参数即节点名称。<br>
>root = etree.Element('root')<br>

##### 2)、为Element对象创建属性
　　属性是以key-value的方式存储的，就像字典一样。<br>
　　可以在创建Element对象时同步创建属性，第二个参数即为属性名和属性值：<br>
>root = etree.Element('root', interesting='totally')<br>

　　也可以使用set方法给已有的Element对象添加属性，两个参数分别为属性名和属性值：<br>
>root.set('hello', 'Huhu')<br>

##### ３)、属性操作
　　属性是以key-value的方式存储的，就像字典一样。直接看例子<br>
>get方法获得某一个属性值<br>
　　　root.get('interesting')<br>
 keys方法获取所有的属性名<br>
　　　sorted(root.keys())<br>
 items方法获取所有的键值对<br>
　　　sorted(root.items())<br>
 attrib属性一次拿到所有的属性及属性值存于字典中<br>
　　　attributes = root.attrib<br>
　　　attributes['good'] = 'Bye'  # 字典的修改影响节点<br>
　　　root.get('good')<br>
##### 4)、文本操作
　　可以使用text和tail属性、或XPath的方式来访问文本内容。<br>
　　一般情况，可以用Element的text属性访问标签的文本。<br>
>root.text <br>
　　
　　XML的标签一般是成对出现的，有开有关，但像HTML则可能出现单一的标签，如下面这段代码中的`<br/>`。<br>
>`<html><body>Text<br/>Tail</body></html>`  <br>

　　Element类提供了tail属性支持单一标签的文本获取。<br>
>html = etree.Element('html')<br>
>body = etree.SubElement(html, 'body')<br>
>body.text = 'Text'<br>
>print(etree.tostring(html))<br>
b'<html><body>Text</body></html>'<br>
>br = etree.SubElement(body, 'br')<br>
>print(etree.tostring(html))<br>
b'<html><body>Text<br/></body></html>'<br>
>br.tail = 'Tail'                       #  tail仅在该标签后面追加文本<br>
>print(etree.tostring(br))<br>
b'<br/>Tail'<br>
>print(etree.tostring(html))<br>
b'<html><body>Text<br/>Tail</body></html>'<br>
>print(etree.tostring(html, method='text'))     #  tostring方法增加method参数，过滤单一标签，输出全部文本<br>
b'TextTail'<br>

　　XPath方式获取文本<br>
>print(html.xpath('string()'))　　　#　 方式一：过滤单一标签，返回文本<br>
TextTail<br>
>texts = html.xpath('//text()')))　　　　# 　方式二：返回列表，以单一标签为分隔<br>

　　方法二获得的列表，每个元素都会带上它所属节点及文本类型信息，如下：<br>
>print(texts[0], texts[0].getparent().tag)　　　　　 #　所属节点名称<br>
Text　body<br>
>print(texts[1], texts[1].getparent().tag)<br>
Tail br<br>
>print(texts[1].is_text)　　　　　　　　　　　　　　　#　 文本类型：是普通文本还是tail文本<br>
False<br>
>print(texts[1].is_tail)<br>
True<br>

##### 10)、Element 的方法介绍：<br>
>append( child )<br>
　　　添加一个新的子节点(可以是 Element 、Comment)到当前 Element 中。<br>
insert( index, elt)<br>
　　　将子元素 elt 插入到指定位置，这个 index 的随意性比较大，如果是正数但是超过了最大值，那么 lxml 会直接将这个元素插到末尾，如果<br>
　　　是负数，但是这个负数所指位置不存在，那么就插到末尾，这个负数的位置计算规则和列表的那个不太一样，不知道正确的规律是什么，但是<br>
　　　经过测试，-n所插的位置，后面有 n (以变化前计算)个元素。如果对 Element 中的子元素执行 insert() 操作，那么子元素位置会按 index<br>
　　　指定的变换。<br>
clear()<br><br>
　　　调用该函数，将移除所有内容 .attrib 将被移除 .text 和 .tail 将被设置为 None 所有的子节点将被删除。<br>
remove( child )<br>
　　　将子节点 child 从Element 中移除 ，如果child 不是 Element 的子节点，将会引发 ValueError 异常。<br>
find( path )<br>
　　　从 Element 的子元素及后代元素中查找第一个符合 path 的 subelement，并且xpath语法只能使用相对路径（以’.//’开头）。如果没有返回 None 。<br>

#### 3、ElementPath：
　　ElementPath需要引入ElementTree类，一个ElementTree对象可理解为一个完整的XML树，每个节点都是一个Element对象。而ElementPath则相当于XML中的XPath。用于搜索和定位Element元素。<br>
　　ElementPath 是 ElementTree 自带的一个 XPath-like 的路径语言，和 XPath 差不太多，主要区别是 ElementPath 能用 {namespace}tag，但是 ElementPath 不能使用值比较和函数。<br>
>findall( path )<br>
　　　返回一个匹配的 Element 的列表。并且xpath语法只能使用相对路径（以’.//’开头）；<br>
findtext( path, default=None )<br>
　　　返回第一个匹配元素的 .text 内容，如果存在匹配，但是没有 .text 内容，那么将返回一个空字符串，如果没有一个匹配的元素，那么将会<br>
　　　返回一个 None ，但是有 default 参数，返回 default 所指定的。<br>
get( key, default=None )<br>
　　　返回字符串形式的 属性 key 的值，没有返回 None 或者 default 指定的。<br>
getchildren()<br>
　　　返回一个包含 Element 子元素的列表。<br>
getiterator( tag=None, \*tags )<br>
　　　返回元素的一个生成器，返回元素类别取决于参数 tag ，生成顺序是in document order (depth first pre-order) 深度优先的先根遍历。如<br>
　　　果没有参数的话，则第一个就是元素本身。如果想使用一个高效的生成器，可以使用 .iter() 。<br>
getroottree()<br>
　　　返回该元素的 ElementTree 。<br>
iter( tag=None, \*tags )<br>
　　　过滤特定标签，生成迭代器。默认情况下，iter() 迭代所有的节点，包括PI(处理指令) 、Comment(注释) 等，如果只想迭代标签元素，可以<br>
   　使用 Element factory 做参数e.iter(tag = etree.Element)。<br>
iterfind( path )<br>
　　　迭代所有匹配 ElementPath 的 Element 。<br>
itertext( tag=None, \*tags, with_tail=True )<br>
　　　迭代 Element 元素的文本内容，with_tail 参数决定是否迭代子元素的 tail 。Element 的tail 不会进行迭代。<br>
iterancestors( tag=None )<br>
　　　如果忽略参数，那么将会迭代所有先辈，加标签名可以只迭代该标签先辈。<br>
iterchildren( reversed=False, tag=None )<br>
　　　通过设置 reversed=True 可以以反向的顺序迭代子元素。<br>
itersiblings( preceding=False )<br>
　　　迭代 Element 之后的兄弟元素，可以通过设置 preceding=True 仅迭代 Element  之前的兄弟元素。<br>
iterdescendants( tag=None )<br>
　　　同 iterancestors()<br>
items()<br>
　　　返回由元素属性的键值所构成的( name, value)元组的列表。<br>
keys()<br>
　　　返回一个没有特定顺序的元素属性名的列表。<br>
set(A, V)<br>
　　　创建或者改变属性 A 的值为 V。<br>
xpath()<br>

#### ５、XPath
　　XPath，全称XML Path Language，即XML路径语言，它是一门在XML文档中查找信息的语言，它最初是用来搜寻XML文档的，但是它同样适用于HTML文档的搜索。<br>
　　XPath的选择功能十分强大，它提供了非常简明的路径选择表达式，另外，它还提供了超过100个内建函数，用于字符串、数值、时间的匹配以及节点、序列的处理等，几乎所有我们想要定位的节点，都可以用XPath来选择。<br>
　　Path 所有的查找都是基于参考定位的，依靠两者之间存在的某种关系，或主从，或平级，这样一步一步的查到。在 XPath 中，有七种类型的节点：元素、属性、文本、命名空间、处理指令、注释以及文档节点（或称为根节点）。XPath 使用路径表达式在 XML 文档中选取节点。节点是通过沿着路径或者 step 来选取的。<br>
　　XPath于1999年11月16日成为W3C标准，它被设计为供XSLT、XPointer以及其他XML解析软件使用，更多的文档可以访问其[官方网站](https://www.w3.org/TR/xpath/all/)。<br>
  下面列出了最有用的**路径表达式**：<br>

|表达式|描述|
|---|---|
|nodename|选取此节点的所有子节点。|
|/|从根节点选取。|
|//|从匹配选择的当前节点选择文档中的节点，而不考虑它们的位置。|
|.|选取当前节点。|
|..|选取当前节点的父节点。|
|@|选取属性。|

　　XPath **通配符**可用来选取未知的 XML 元素。<br>

|通配符|描述|
|---|---|
|\*|匹配任何元素节点。|
|@\*|匹配任何属性节点。|
|node()|匹配任何类型的节点。|

通过在路径表达式中使用“|”运算符，可以选取若干个路径，“|”两边的表达式是互相独立的，并非是一个表达式下。<br>
除了上面介绍的还有两个概念需要知道：**轴(Axes) 、谓语**。<br>
通过以上介绍我们基本可以进行选取节点了，但是有可能选的节点有点多，那么这时候就需要用谓语了。谓语就是用来做过滤的，过滤条件均写在[]中。<br>
**谓语**的过滤条件可写内容有很多：运算符表达式、数字、函数以及文字。<br>

|路径表达式|结果|
|---|---|
|/bookstore/book[1]|选取属于 bookstore 子元素的第一个 book 元素。|
|/bookstore/book[last()]|选取属于 bookstore 子元素的最后一个 book 元素。|
|/bookstore/book[last()-1]|选取属于 bookstore 子元素的倒数第二个 book 元素。|
|/bookstore/book[position()<3]|选取最前面的两个属于 bookstore 元素的子元素的 book 元素。|
|//title[@lang]|选取所有拥有名为 lang 的属性的 title 元素。|
|//title[@lang='eng']|选取所有 title 元素，且这些元素拥有值为 eng 的 lang 属性。|
|/bookstore/book[price>35.00]|选取 bookstore 元素的所有 book 元素，且其中的 price 元素的值须大于 35.00。|
|/bookstore/book[price>35.00]/title|选取 bookstore 元素中的 book 元素的所有 title 元素，且其中的 price 元素的值须大于 35.00。|

**Xpath 的轴**定义的是相对于当前节点具有某种相同特征的节点集。<br>

|轴名称|结果|
|---|---|
|ancestor|选取当前节点的所有先辈（父、祖父等）。|
|ancestor-or-self|选取当前节点的所有先辈（父、祖父等）以及当前节点本身。|
|attribute|选取当前节点的所有属性。|
|child|选取当前节点的所有子元素。|
|descendant|选取当前节点的所有后代元素（子、孙等）。|
|descendant-or-self|选取当前节点的所有后代元素（子、孙等）以及当前节点本身。|
|following|选取文档中当前节点的结束标签之后的所有节点。|
|namespace|选取当前节点的所有命名空间节点。|
|parent|选取当前节点的父节点。|
|preceding|选取文档中当前节点的开始标签之前的所有节点。|
|preceding-sibling|选取当前节点之前的所有同级节点。|
|self|选取当前节点。|

　　注意：每一个step都是由轴名称::NodeTest[谓语]构成。比如:child::*/child::price中child::*和child::price都是一个step。<br>

**xpath其他用法**：<br>
>position定位<br>
　　　tree.xpath('//\*[@id="testid"]/ol/li[position()=2]/text()')[0] <br>
条件判断<br>
　　　print tree.xpath(u'//h2[text()="这里是个小标题"]/text()')[0]<br>
count：统计<br>
　　　print tree.xpath('count(//li[@data])')　　　　　　　#　节点统计<br>
concat：字符串连接<br>
　　　print tree.xpath('concat(//li[@data="one"]/text(),//li[@data="three"]/text())')<br>
string：解析当前节点下的字符<br>
　　　print tree.xpath('string(//li)') <br>
注意：tring只能解析匹配到的第一个节点下的值，也就是作用于list时只匹配第一个<br>
local-name：解析节点名称<br>
　　　print tree.xpath('local-name(//\*[@id="testid"])')　　　　 #　local-name解析节点名称<br>
contains(string1,string2)：如果 string1 包含 string2，则返回 true，否则返回 false<br>
　　　tree.xpath('//h3[contains(text(),"H3")]/a/text()')[0]　　　　　　　　 #　使用字符内容来辅助定位<br>
组合拳：匹配带有href属性的a标签的先辈节点中的div，其兄弟节点中前一个div节点下ul下li中text属性包含“务”字的节点的值<br>
　　　print tree.xpath(u'//a[@href]/ancestor::div/preceding::div/ul/li[contains(text(),"务")]/text()')[0] <br>
注意：兄弟节点后一个节点可以使用：following-sibling<br>
not：布尔值（否）<br>
　　　print tree.xpath('count(//li[not(@data)])')　　　　　　　　 #　不包含data属性的li标签统计<br>
string-length：返回指定字符串的长度，如：string-length函数+local-name函数定位节点名长度小于2的元素<br>
　　　print tree.xpath('//\*[string-length(local-name())<2]/text()')[0] <br>
组合拳：contains函数+local-name函数定位节点名包含di的元素<br>
　　　print tree.xpath('//div[@id="testid"]/following::div[contains(local-name(),"di")]') <br>
or：多条件匹配<br>
　　　print tree.xpath('//li[@data="one" or @code="84"]/text()')　　　　　　 #　or匹配多个条件<br>
也可使用|<br>
　　　print tree.xpath('//li[@data="one"]/text() | //li[@code="84"]/text()') 　　　　　#　|匹配多个条件<br>
组合拳：floor + div除法 + ceiling；如：position定位+last+div除法，选取中间两个<br>
　　　tree.xpath('//div[@id="go"]/ul/li[position()=floor(last() div 2+0.5) or position()=ceiling(last() div 2+0.5)]/text()') <br>
组合拳：隔行定位：position+mod取余<br>
　　　tree.xpath('//div[@id="go"]/ul/li[position()=((position() mod 2)=0)]/text()') 　　　#　position+取余运算隔行定位<br>
starts-with：以．．．开始<br>
　　　print tree.xpath('//li[starts-with(@code,"8")]/text()')[0]　　　　　#　starts-with定位属性值以8开头的li元素<br>
数值比较：小于（<），<br>
　　　print tree.xpath('//li[@code<200]/text()')　　　　　#　所有li的code属性小于200的节点<br>
div：对某两个节点的属性值做除法<br>
　　　print tree.xpath('//div[@id="testid"]/ul/li[3]/@code div //div[@id="testid"]/ul/li[1]/@code')<br>
组合拳：根据节点下的某一节点数量定位<br>
　　　print tree.xpath('//ul[count(li)>5]/li/text()')　　　　　#　选取所有ul下li节点数大于5的ul节点<br>
将对象还原为字符串<br>
　　　s = tree.xpath('//\*[@id="testid"]')[0] 　　　　　　#　使用xpath定位一个节点<br>
　　　s2 = etree.tostring(s).decode('utf-8') 　　　　　　#　还原这个对象为bytes格式，在decode为字符串<br>

### 三、lxml.html

　　lxml.html 是一个专门用来处理 HTML 文档的，虽然他依旧是基于 HTML parser 的，但是她提供了很多有针对性的 API 。<br>
首先在解析 HTML 文本上增加了几个方法：<br>
>document_fromstring(string)<br>
fragment_fromstring(string, create_parent=False)<br>
fragments_fromstring(string)<br>

其次对于 HTML 元素，除了之前的方法之外又加了一些：<br>
>.drop_tree()<br>
　　　删除该元素及其所有子元素，就像 el.getparent().remove(el)，但是不会删除该元素的 .tail ，.tail 将会与前一个元素合并。<br>
.drop_tag()<br>
　　　删除该标签，但是会保留该元素的子元素与文本。<br>
.find_class(class_name)<br>
　　　返回一个含有 class_name CSS 的元素列表。<br>
.find_rel_links(rel)<br>
　　　返回一个<a \rel="{rel}">的元素列表<br>
.get_element_by_id(id, default=None)<br>
　　　通过 id 查找元素，如果没有返回默认，假如文档不合规出现多个同 id 的元素，仅返回第一个。<br>
.text_content()<br>
　　　返回元素的文本内容，包括其子代的文本内容。<br>
.cssselect(expr)<br>
　　　通过 CSS 选择器选择元素。<br>
.set(key, value=None)<br>
　　　设置元素属性，如果 value 没有给或者是 None ，那么将创建一个 boolean attribute (属性只要出现即表示真，无论 value 是什么，哪怕是个空串) 

最后，lxml.html.soupparser 可以引入 BeautifulSoup ，甚至可以仅使用 BeautifulSoup 的编码检测模块 UnicodeDammit 。可以使用 lxml.html.open_in_browser(lxml_doc) 直接在浏览器中显示文档，甚至可以对文档的表单进行提交操作。<br>
