### 一、selenium 等待响应完成（元素加载完成）：
　　selenium是自动运行浏览器的一种行为，而浏览器的加载是需要时间的，我们在进行操作之前肯定必须要等到被操作的元素加载出来后才行，所以我们就需要用到等待的操作。<br>

**等待的方式分三种：**<br>
1、强制等待：<br>
>time.sleep（）<br>


　　强制等待一个固定时长，时间到了就继续执行下面语句，如：sleep（3），3s过后执行后面的代码。这种等待方式比较死板，效率低，并且不能保证在规定的时间内元素就被加载出来了。<br>

2、隐性等待：<br>
定义：<br>
WebDriver提供了implicitly_wait()方法来实现隐式等待，默认设置为0，用法相对简单。<br>
语法：<br>
>driver.implicitly_wait(10)<br>

注意：<br>
implicitly_wait()默认参数的单位为秒，延时等待是在设定好的时间内，不断进行元素定位（轮询），定位到了就继续执行，定位不到则在设定的时间后抛出异常。<br>
一旦设置隐式等待时间，就会作用于这个WebDriver实例的整个生命周期（对所有的元素查找都生效），设置隐式等待时间后，Webdriver会在一定时间内持续检测和搜寻DOM，以便于查找一个或多个不是立即加载成功并可用的元素。<br>
用例：<br>
>from selenium import webdriver<br>
from selenium.common.exceptions import NoSuchElementException<br><br>
driver = webdriver.Chrome()　　　　# 隐式等待设置为10秒，如果等2秒就能加载出来则直接进行下面操作<br>
driver.implicitly_wait(10)<br>
driver.get("https://www.baidu.com")<br>
try:<br>
    driver.find_element_by_id('kw22').send_keys('selenium')<br>
except NoSuchElementException as e:<br>
    print(e)<br>

隐式等待是通过一定时长来等待页面上某个元素的加载完成，webdriver提供了implicitly_wait()方法来实现隐式等待，默认为0，超时抛出 **NoSuchElementException 异常**。<br>
语法：driver.implicitly_wait()，时间为秒。<br>
假设等待时间为8秒，这8秒不是一个固定时间，不会影响脚本的执行效率，它不针对页面上的某个元素进行等待。当脚本执行到某个元素定位时，如果元素可以定位到，则继续执行，如果定位不到，则会轮询进行定位，假设在3秒时，元素定位到，则继续执行。如果超过8秒，则抛出异常。<br>


3、显性等待：<br>
参考：[元素等待机制](https://www.cnblogs.com/simple-free/p/8458361.html)<br>
定义：<br>
使webdriver等待某个条件成立时继续执行，否则达到最大时长时抛出异常。<br>
WebDriver提供了WebDriverWait类和expected_conditions模块来实现显式等待。相比隐式等待，显示等待更加智能。显示等待就是设置一个前置条件，在等待时间内，每隔一段时间检查一次前置条件是否满足，满足则执行下一步，超时则报TimeoutException异常。<br>
WebdriverWait类是由 WebDriver 提供的等待方法。在设置时间内，默认每隔一段时间检测一次当前页面元素是否存在，如果超过设置时间检测不到则会抛出异常。<br>
语法：<br>
>WebDriverWait(driver, timeout, poll_frequency=0.5, ignored_exceptions=None)<br>

参数：<br>
>driver:　　　　　　　　　浏览器驱动<br>
timeout:　　　　　　　　超时时间，单位为秒<br>
poll_frequency：　　　　检测的间隔时间，默认为0.5s<br>
ignored_exceptions：　　超时后抛得异常，默认情况下抛 **NoSuchElementException 异常**。<br>

注意：<br>
WebDriverWait()一般由until()或until_not()配合使用。<br>
用例：<br>
>from selenium import webdriver<br>
from selenium.webdriver.common.by import By<br>
**from selenium.webdriver.support.ui import WebDriverWait**<br>
from selenium.webdriver.support import expected_conditions as EC<br><br>
driver = webdriver.Chrome()<br>
driver.get("https://www.baidu.com")<br>
\# 显式等待 WebDriverWait(driver,5,0.5) 每隔0.5秒刷新一次，如果5秒内能定位到这个元素，则通过，否则抛出 **NoSuchElementException 异常**<br>
element = WebDriverWait(driver,5,0.5).until(EC.presence_of_element_located((By.ID,'kw')))<br>
element.send_keys('selenium')<br>
driver.quit()<br>

WebDriverWait,配合该类的until()和until_not()方法，表示程序每隔x秒去判断一下指定的元素是否加载完，加载完了就执行下一步，否则继续每隔x秒去判断，到指定时间截止。如果超时就会抛出异常。<br>
until方法 和 until_not方法：<br>
until（method, message=''）：method指要执行的方法（等待时间内每隔一段时间，执行一次），直到返回值为true，超时则报TimeoutException异常，message将传入异常（message参数可不填）。<br>
until_not（method, message=''）：直到返回值为false，其他和until相同。<br>
expected_conditions模块提供了多种定义好的前置条件，需要配合WebDriverWait使用:<br>
>**1、**element_to_be_clickable(locator)　　#　等待查找的元素可见并且可用，以便可以点击，返回定位到的元素<br>
参数：locator，指元组（By,locator）#例如：WebDriverWait(driver,10).until(expected_conditions.element_to_be_clickable((By.NAME,'11')))　　#　下面的都是以这种方式，只是前置条件不同，传的参数也可能不同。<br>
**2、**element_to_be_selected(locator)　　　　等待直到元素被选中<br>
**3、**invisibility_of_element_located(locator)　　　　等待一个元素在DOM中不可见 或不存在<br>
**4、**presence_of_all_elements_located(locator)　　　　所有元素加载出；等待至少有一个定位器查找的元素出现在网页中，返回一组元素<br>
**5、**presence_of_element_located(locator)　　　　等待定位器查找的元素出现在网页中，或者可以在DOM中找到，返回一个被定位到的元素<br>
**6、**text_to_be_present_in_element(locator,text)　　　　等待元素能被定位，并且带有指定的文本信息<br>
参数：text，指定的文本<br>
**7、**title_contains(title)　　　　等待网页标题包含指定的字符串，成功时返回True，否则返回false<br>
参数：title，指要校验标题包含的字符串<br>
**8、**title_is(title)　　　　等待网页标题是某内容，成功时返回True，否则返回false<br>
参数：title，指要校验的标题<br>
**9、**visibility_of(element)　　　　等待元素出现在DOM中，是可见的，并且宽和高都大于0，变为可见的，将返回一个元素（同一个）<br>
参数：element，指一个元素对象<br>
**10、**visibility_of_element_located(locator)　　　　 等待元素出现在DOM中，是可见的，并且宽和高都大于0，变为可见的，将返回一个元素<br>
**11、**alert_is_present()　　　　判断是否存在警告窗口<br>
**12、**text_to_be_present_in_element_value　　　　  某个元素值包含某文字<br>
**13、**frame_to_be_available_and_switch_to_it frame　　　　 加载并切换<br>
**14、**staleness_of　　　　  判断一个元素是否仍在DOM，可判断页面是否已经刷新<br>
**15、**element_located_to_be_selected 　　　　 元素可选择，传入定位元组<br>
**16、**element_selection_state_to_be 　　　　 传入元素对象以及状态，相等返回True，否则返回False<br>
**17、**element_located_selection_state_to_be 　　　　 传入定位元组以及状态，相等返回True，否则返回False<br>

**浏览器动作、拖拽操作：**<br>
>**from selenium.webdriver import ActionChains**　　　#引入动作链<br>
>#　定义一个ActionChains类的对象，参数是浏览器对象<br>
>#　ActionChains类，需要一个参数：浏览器对象。当调用相关鼠标操做函数的时候，这些事情并不会立刻执行，而是将所有的鼠标事件都放入到了一个队列当中，只有执行perform函数的时候，所有的鼠标事件才会被执行。<br>

source = browser.find_element_by_css_selector('#draggable')   #找到被拖拽对象<br>
target = browser.find_element_by_css_selector('#droppable')   #找到目标<br>
actions.drag_and_drop(source, target)　　　#　拖拽到某个元素然后松开<br>
actions.drap_and_drop_by_offset(source,x,y)　　　#　拖拽到距离某个坐标点多少的位置然后松开<br>
action.click_and_hold(source).move_to_element(target).release()<br>
actions.perform()   #执行动作<br>
