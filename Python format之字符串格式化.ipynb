## Python format之字符串格式化

　　相对基本格式化输出采用‘%’的方法，format()功能更强大，该函数把字符串当成一个模板，通过传入的参数进行格式化，并且使用大括号‘{}’作为特殊字符代替‘%’<br>
使用方法由两种：b.format(a)和format(a,b)。<br>
### 1、基本用法
　　（1）不带编号，即“{}”<br>
　　（2）带数字编号，可调换顺序，即“{1}”、“{2}”<br>
　　（3）带关键字，即“{a}”、“{tom}”<br>
>print('{} {}'.format('hello','world'))  # 不带字段<br>
 hello world<br>
 print('{0} {1}'.format('hello','world'))  # 带数字编号<br>
 hello world<br>
 print('{0} {1} {0}'.format('hello','world'))  # 打乱顺序<br>
 hello world hello<br>
 print('{1} {1} {0}'.format('hello','world'))<br>
 world world hello<br>
 print('{a} {tom} {a}'.format(tom='hello',a='world'))  # 带关键字<br>
 world hello world<br>
 
 ### 2、进阶用法<br>
　　（1）< （默认）左对齐、> 右对齐、^ 中间对齐、= （只用于数字）在小数点后进行补齐<br>
　　（2）取位数“{:4s}”、"{:.2f}"等<br>
>print('{} and {}'.format('hello','world'))  # 默认左对齐<br>
 hello and world<br>
 print('{:10s} and {:>10s}'.format('hello','world'))  # 取10位左对齐，取10位右对齐<br>
 hello      and      world<br>
 print('{:^10s} and {:^10s}'.format('hello','world'))  # 取10位中间对齐<br>
 hello    and   world   <br>
 print('{} is {:.2f}'.format(1.123,1.123))  # 取2位小数<br>
 1.123 is 1.12<br>
 print('{0} is {0:>10.2f}'.format(1.123))  # 取2位小数，右对齐，取10位<br>
 1.123 is       1.12<br>
 
 ### 3、多种格式化<br>
　　（1）'b' - 二进制。将数字以2为基数进行输出。<br>
　　（2）'c' - 字符。在打印之前将整数转换成对应的Unicode字符串。<br>
　　（3）'d' - 十进制整数。将数字以10为基数进行输出。<br>
　　（4）'o' - 八进制。将数字以8为基数进行输出。<br>
　　（5）'x' - 十六进制。将数字以16为基数进行输出，9以上的位数用小写字母。<br>
　　（6）'e' - 幂符号。用科学计数法打印数字。用'e'表示幂。<br>
　　（7）'g' - 一般格式。将数值以fixed-point格式输出。当数值特别大的时候，用幂形式打印。<br>
　　（8）'n' - 数字。当值为整数时和'd'相同，值为浮点数时和'g'相同。不同的是它会根据区域设置插入数字分隔符。<br>
　　（9）'%' - 百分数。将数值乘以100然后以fixed-point('f')格式打印，值后面会有一个百分号。<br>
>print('{0:b}'.format(3))<br>
 11<br>
 print('{:c}'.format(20))<br>
 (结果为空格)<br>
 print('{:d}'.format(20))<br>
 20<br>
 print('{:o}'.format(20))<br>
 24<br>
 print('{:x}'.format(20))<br>
 14<br>
 print('{:e}'.format(20))<br>
 2.000000e+01<br>
 print('{:g}'.format(20.1))<br>
 20.1<br>
 print('{:f}'.format(20))<br>
 20.000000<br>
 print('{:n}'.format(20))<br>
 20<br>
 print('{:%}'.format(20))<br>
 2000.000000%<br>

### 4、通过属性匹配参数
>c = 3-5j<br>
('The complex number {0} is formed from the real part {0.real} '<br>
...  'and the imaginary part {0.imag}.').format(c)<br>
'The complex number (3-5j) is formed from the real part 3.0 and the imaginary part -5.0.'<br>
class Point:<br>
... def __init__(self, x, y):<br>
...    　　self.x, self.y = x, y<br>
... def __str__(self):<br>
...    　　return 'Point({self.x}, {self.y})'.format(self=self)<br>
str(Point(4, 2))<br>
'Point(4, 2)'<br>

### 5、通过下标或key匹配参数
>coord = (3, 5)<br>
'X: {0[0]};  Y: {0[1]}'.format(coord)<br>
'X: 3;  Y: 5'<br>
a = {'a': 'test_a', 'b': 'test_b'}<br>
'X: {0[a]};  Y: {0[b]}'.format(a)<br>
'X: test_a;  Y: test_b'<br>

### 6、占位符%s和%r<br>
replacement_field ::= "{" [field_name] ["!" conversion] [":" format_spec] "}"<br>
conversion ::= "r" | "s" | "a"<br>
这里只有三个转换符号，用"!"开头。<br>
"!r"对应 repr()；"!s"对应 str(); "!a"对应ascii()。<br>
>"repr() shows quotes: {!r}; str() doesn't: {!s}".format('test1', 'test2')<br>
"repr() shows quotes: 'test1'; str() doesn't: test2"  # 输出结果是一个带引号，一个不带

### 7、设置长度及对齐
>'{:<30}'.format('left aligned')  　　# 左对齐<br>
'left aligned  <br>                '
'{:>30}'.format('right aligned') 　　 # 右对齐<br>
'                 right aligned'<br>
'{:^30}'.format('centered') 　　　　 # 中间对齐<br>
'           centered           '<br>
'{:*^30}'.format('centered')　　　　  # 使用“*”填充<br>
'***********centered***********'<br>
'{:0=30}'.format(11)  # 还有“=”只能应用于数字，这种方法可用“>”代替<br>
'000000000000000000000000000011'<br>

### 8、%+f, %-f, 和 % f的用法<br>
> '{:+f}; {:+f}'.format(3.14, -3.14)  # 总是显示符号<br>
'+3.140000; -3.140000'<br>
'{: f}; {: f}'.format(3.14, -3.14)  # 若是+数，则在前面留空格<br>
' 3.140000; -3.140000'<br>
'{:-f}; {:-f}'.format(3.14, -3.14)  # -数时显示-，与'{:f}; {:f}'一致<br>
'3.140000; -3.140000'<br>

### 9、按2进制、10进制、16进制显示
>\# format also supports binary numbers<br>
"int: {0:d};  hex: {0:x};  oct: {0:o};  bin: {0:b}".format(42)<br>
'int: 42;  hex: 2a;  oct: 52;  bin: 101010'<br>
>\# with 0x, 0o, or 0b as prefix:<br>
"int: {0:d};  hex: {0:#x};  oct: {0:#o};  bin: {0:#b}".format(42)  # 在前面加“#”，则带进制前缀<br>
'int: 42;  hex: 0x2a;  oct: 0o52;  bin: 0b101010'<br>

### 10、用“，”分隔数字，每一千进位
> '{:,}'.format(1234567890)<br>
'1,234,567,890'<br>

### 11、百分数%

total = 22<br>
'Correct answers: {:.2%}'.format(points/total)<br>
'Correct answers: 86.36%'<br>

### 14、另，可在字符串前加f以达到格式化的目的，在{}里加入对象，此为format的另一种形式：

age = 18<br>
sex = 'man'<br>

print(f'My salary is {salary:10.3f}')<br>
\# 结果<br>
my name is Jack.<br>
I am ****18**** years old.<br>
I am a man<br>
My salary is   9999.990<br>
