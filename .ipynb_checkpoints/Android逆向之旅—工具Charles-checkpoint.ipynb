## 一、介绍
　　Charles是一个运行在PC上web代理(HTTP代理/ HTTP监视器)。web浏览器(或其他Internet应用程序)可通过Charles访问Internet，然后Charles能够为您记录和显示发送和接收的所有数据。它允许一个开发者查看所有连接互联网的HTTP通信，这些包括request, response和HTTP headers （包含cookies与caching信息）。
　　charles还是比较智能的，charles运行时会在本地建立一个代理服务，默认地址：127.0.0.1:8888，当我们启动charles的时候会自动给我们的浏览器设置代理，浏览器的请求我们就可以在charles面板看到了，但是此时只能捕获到http的请求，https的请求需要我们进一步配置。如果启动charles后并没有捕获到请求，请检查你的浏览器是否使用了代理软件，这里是个大坑，一定要注意。
　　Charles不仅可以帮助我们抓取服务器返回的数据，还可以修改请求参数，修改响应体。其主要功能有：  
+ 支持SSL代理。可以截取分析SSL的请求。
+ 支持流量控制。可以模拟慢速网络以及等待时间（latency）较长的请求。
+ 支持AJAX调试。可以自动将json或xml数据格式化，方便查看。
+ 支持AMF调试。可以将Flash Remoting 或 Flex Remoting信息格式化，方便查看。
+ 支持重发网络请求，方便后端调试。
+ 支持修改网络请求参数。
+ 支持网络请求的截获并动态修改。
+ 检查HTML，CSS和RSS内容是否符合W3C标准。

## 二、安装
　　下载地址：[https://www.charlesproxy.com/](https://www.charlesproxy.com/)  

　　**Charles 激活：**
  + 1、注册码生成：[Charles 激活码计算器](https://www.zzzmode.com/mytools/charles/)  
  + 2、激活Charles: 在Help->Registered输入key和生成的激活码即可。

## 三、PC抓包
### 3.1、安装Charles证书
　　点击顶部菜单栏【Help】-> 选择【SSL Proxying】-> 点击【install Charles Root Certificate 】-> 弹窗点击【安装证书】-> 选择'本地计算机'点【下一步】-> 选择'将所有的证书都放入下列存储'点击【浏览】-> 选择'受信任的根证书颁发机构' 点击【确定】-> 点击【下一页】 -> 点击【完成】&【确定】
  ![证书安装图](./images/Charles-证书安装.jpg)  
### 3.2、SSL代理设置
　　点击charles菜单栏【Proxy】-> 选择【SSL Proxying Settings…】-> 在弹出选项卡中，勾选【Enable SSL Proxying】-> 点击【add】-> 在Host输入【\*】表示接收任何主机 -> 在Prot输入【443】-> 最后点击【ok】保存***
  ![SSL代理设置图](./images/Charles-SSL代理设置.jpg) 

## 四、移动端抓包

### 4.1、PC端配置
### 4.1.1、启动Charles，配置SSL代理

### 4.2.2、Charles Proxy配置
　　点击【Proxy】–>【Proxy Settings】我们在设置下端口号，选择 【HTTP Proxy Mode】，确定即可。
### 4.2.2、获取PC IP地址
　　点击菜单栏【Help】->【local IP Address】获取下ip地址，也可以直接通过电脑自带的CMD命令来获取电脑IP【ipconfig】
  

经过上面的设置之后，我们就可以对手机的网络请求进行抓包了，但是如果请求的是 HTTPS，抓到的包无法查看，需要电脑和手机安装证书 (Help -> SSL Proxying)

### 4.2、移动端配置
　　Charles配置代理抓包与Fiddler相似，都是配置移动端的代理为主机，同时安装证书到移动端系统目录。弊端也是一样，当APP配置不走代理，或一些需要使用VPN访问的情况，都无法进行抓包。  

### 4.2.1、移动端配置代理
　　1、移动端设配需与PC设备处于同一网段内。  
　　2、Android 设备长按连接的wifi(或编辑连接的wifi,点击高级选项)，代理配置选择【手动】 -> 代理服务器填：PC IP地址 -> 代理服务器端口：抓包工具的代理配置端口，一般为：8888 -> 保存即可。   
　　注意：设置好之后，Charles 会弹出一个框，提示是否允许手机进行连接，接受即可。如果手滑，点了拒绝【Deny】，也可以在这里：【Proxy】 -> 【Access Control Settings】 进行添加你的手机IP。　　
　　此时，简单的HTTP请求的数据包已经可以抓取查看了。 
  
### 4.2.2、移动端安装证书
　　1、Charles 点击【help】-> 选择【SSL Proxying】-> 选择【Install Charles Root Certificate On a Mobile Device or Remote Browser】，在移动端安装证书。（从弹窗提示可见：要求移动端配置PC端Charles代理，并到chls.pro/ssl 下载证书安装。有的手机可能会提示：无法安装该证书，因为无法读取证书文件。解决方法：不要使用系统的浏览器下载，使用第三方浏览器下载 baidu、UC等）   
　　2、证书下载完成后，在手机上安装该证书，流程如下(小米手机为例)：打开手机 设置 -> 更多设置 -> 系统安全 -> 从SD卡安装 , 然后输入你自定义的证书名称即可(有些手机可能需要输入密码凭证)。
　　3、**Android 7.0 以后，处于安全的考虑，系统默认只会信任系统(System)级别的证书,不再信任用户添加的证书(User),此时证书需要被导入为系统证书，才能正常验证通过**，此时参考:[Android逆向之旅—Android抓包与工具证书导入为系统证书](./Android逆向之旅—Android抓包与工具证书导入为系统证书.ipynb) 使用Magisk的Move Certificates模块, 把用户证书迁移为系统证书。  
>系统证书路径 /etc/security/cacerts   
用户证书路径 /data/misc/user/0/cacerts-added

　　4、关于android开发配置：  
　　一般我们只在 Debug版本 允许抓包，正式版本不应该允许抓包。可以 AndroidManifest.xml 文件中配置安全策略：
  
```xml
<application android:networkSecurityConfig="@xml/network_security_config" >
</application>

```
然后在res目录下新建xml文件夹，然后在xml文件夹中新建 network_security_config.xml:
```xml
<?xml version="1.0" encoding="utf-8"?>
<network-security-config>
    <debug-overrides>
        <trust-anchors>
            <certificates src="system" />
			<!-- debug版本情况下，信任用户添加的证书 -->
            <certificates src="user" />
        </trust-anchors>
    </debug-overrides>
    <base-config cleartextTrafficPermitted="true">
        <trust-anchors>
            <certificates src="system" />
        </trust-anchors>
    </base-config>
</network-security-config>

```

### 4.3、Charles + Postorn 抓包
　　与Fiddler不同的是，Charles支持socket代理，这样也就可以与Postern等VPN工具配合，避过APP不走代理的坑了。 此时Charles不在直接监听APP,而是监听到了VPN(Postern)上 ,这样VPN将APP的发包转发到Charles上，Charles也就可以对app进行抓包了。另外，Charles + Postern 的方案，不需要在手机上配置wifi代理，要相对操作简单一些。   
  
#### 4.3.1、去除windows proxy 选项
　　菜单栏 -> Proxy -> 去除 Windows Proxy 前对勾。  
#### 4.3.2、Charles 设置socket代理
　　菜单栏 -> Proxy -> Proxy Settings... -> 勾选 Enable SOCKET proxy()    
#### 4.3.3、Postern配置
　　Postern并不是抓包工具，它是一个代理工具。它可以将http请求转为socket，并且包转发到Charles上，这样就可以抓到更多的包。   
　　打开移动端Postern -> 配置代理（设为PC的IP以及Charles配置的Socks端口）-> 配置规则（添加规则的代理为上一步配置的代理）-> 打开VPN。   
　　注意： charles有弹窗，需要点击Allow。   



## 五、弱网测试
　　Charles配置弱网设置：
+ 1、点击【Proxy】–>【Throttle Setting】
+ 2、勾选上【Enable Throttle】，也可以勾选【only for selected host 】 设置一个指定主机访问进行网络限制，不影响其他软件；
+ 3、在【Throttle preset】选择常见预设的网络情况，也可以在【Bandwidth】自己设置网络的上传和下载速率
　　![Charles弱网配置](./images/Charles-弱网配置.png)  

## 六、Charles过滤
　　通常情况下，我们需要对网络请求进行过滤，只监控向指定目录服务器上发送的请求。对于这种需求，以下几种办法：
### 6.1、方法一
　　在主界面的中部的 Filter 栏中填入需要过滤出来的关键字。例如我们的服务器的地址是：http://ceshi.com, 那么只需要在 Filter 栏中填入 ceshi 即可。

### 6.2、方法二
　　在 Charles 的菜单栏选择 “Proxy”->”Recording Settings”，然后选择 Include 栏，选择添加一个项目，然后填入需要监控的协议，主机地址，端口号。这样就可以只截取目标网站的封包了。
### 6.3、方法三
　　1、在SSL代理设置的【SSL Proxying Settings】处，添加 Include 或 Exclude 项，添加或排除抓取的域名。  
　　2、直接选择开启，在首页Structure栏目下选择想要捕获https的域名，然后右键选择Enable SSL Proxying开启即可。使用这种方式会自动在SSL Proxying栏目下添加该域名。  
## 七、断点调试
　　可以使用Charles在某个网络请求的时候进行断点， 这样可以精确的定位某个网络请求：
![断点调试](./images/Charles-断点调试.png)

　　除了上面的手动添加需要断点的URL，如果这个URL已经请求过，可以对该请求单击右键，然后点击 Breakpoints。

　　通过断点功能，可以修改 Request 参数 和 Response 返回值。需要注意的是修改完 Request 参数需要点击 Excute 按钮，请求成功后，才能修改 Response 返回值。
　　![断点调试修改包](./images/Charles-断点调试修改包.png)

## 八、修改HTTP包
+ 1、选择请求会话右键勾选【Breakpoints】这是我们会发现工具栏有个六边形按钮已经变为红色
+ 2、设置断点，点击菜单栏【proxy】勾选【Breakoint Setting …】这时断点设置弹窗；
+ 3、在弹出的断点设置弹窗，双击想打断点的接口，进行断点编辑；
+ 4、编辑断点，如果修改Request数据，就将Request勾选，如果修改Response数据，就将Response勾选，当然你也可以都勾选；
+ 5、在手机上重新进行请求，Charles 会自动进行拦截并跳转到请求数据修改页面；

　　Charles 可以通过 Compose 功能修改请求的 URL、Method、Header、Parameter等。  
　　修改方式：对请求单击右键，然后点击 compose 按钮。

## 十、Rewrite重写功能
　　Rewrite重写功能是全局设置的，并不是对某个请求单击右键的方式临时操作下， Rewrite可以：
+ 修改请求头(Add/Modify/Remove Header)
+ 修改Host/Path/URL
+ 修改请求参数(Add/Modify/Remove Query-Parameter)
+ 修改响应状态(Response Status)
+ 修改Body
　　Rewrite设置路径： Tools -> Rewrite
![设置路径](./images/Charles-Rewrite设置路径.png)
　　Rewrite重写顾名思义就是替换，所以是基于某个固定的值替换成某个值，添加规则的时候，支持正则表达式：
![Rewrite正则替换](./images/Charles-正则替换Rewrite.png)  


## 十一、Map映射功能
　　Map映射包含两种一种是 Map Remote，另一种是 Map Local。设置路径：Tools -> Map Local/Remote  
　　Map Remote 顾名思义就是把匹配的请求映射到另一个远程URL：
![Map remote](./images/Charles-Map remote.png)  
　　Map Local 是将匹配的请求映射到本地文件：
![Map local](./images/Charles- Map local.png)


## 十二、Repeat 重试功能
　　Repeat 功能有两种：一种是简单的重试一次(Repeat)；还有一种 Repeat advances 可以设置重试的次数和并发的数量。
　　重试一次主要用于是和服务端联调，如果接口有问题，服务端开发人员可能需要在后台代码打断点的方式来排查问题。这个时候，我们不需要在app中重新重复上次的请求(可能需要填写字段比较多或者界面链路比较长)，只需要在Charles中对该请求单击右键，然后 Repeat 即可。  
　　Repeat advances 可以来对服务器进行压力测试，或者批量提交写数据等，如：
![Charles-批量重试](./images/Charles-批量重试.png)
