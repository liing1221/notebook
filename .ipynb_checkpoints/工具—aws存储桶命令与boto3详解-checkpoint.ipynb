## 一、简介
　　aws 是亚马逊发布的一套命令行工具，用于管理操作aws上的资源。它的工作原理类似于boto，可以使用sudo easy_install awscli或sudo pip install awscli安装。  


## 二、aws 命令
　　aws命令安装
```shell
pip install awscli
```
　　aws 命令配置，此处用到远程登录的相关s3配置:
```shell
aws configure   --profile product　　　　# initailize configuration,input the aws access key
# 输入如下输入项后，回车配置成功
AWS Access Key ID [None]:
AWS Secret Access Key [None]:
Default region name [None]:
Default output format [None]:
```
　　配置完后，就会用户home目录下生成.aws目录，里面的文件会记录我们的登录凭证。   
   
　　aws安装配置后，可以使用linux命令行的一些命令执行s3存储桶的操作，只是在linux命令前需加上 aws s3, 如下：  
```shell
aws s3 ls(ls命令) --region cn-northwest-1 s3://adhoc/test
```
　　登录aws：
```shell
aws s3 ls --region cn-northwest-1
```
　　查看文件夹adhoc/test
```shell
aws s3 ls --region cn-northwest-1 s3://adhoc/test/
```
　　aws创建名为nlp的bucket
```shell
aws s3 mb --region cn-northwest-1 s3://nlp
```
　　aws删除名为nlp的bucket(此处有错)
```shell
aws s3 mb --region cn-northwest-1 s3://nlp
```

　　aws上传文件夹以及文件夹下的文件
```shell
aws s3 cp ./data/ --region cn-northwest-1 s3://nlp/data/ --recursive
```
　　aws下载文件或文件夹下的所有文件 --region可省
```shell
aws s3 cp --region cn-northwest-1 s3://nlp/test.txt --recursive　　　　# 下载文件: --recursive 用于复制下载多个文件 
aws s3 cp --region cn-northwest-1 s3://nlp/data/ ./data/ --recursive　　　　# 下载文件夹下面的所有文件
aws s3 sync s3://nlp/data/ ./data    # sync 命令将复制整个目录，它只会复制新的/修改过得文件  

```
　　删除aws上文件
```shell
aws s3 rm s3://nlp/test.txt
```
　　删除aws上文件夹以及文件夹下的文件
```
aws s3 rm s3://nlp/test —recursive
```
　　以下也是s3的下载命令
> aws s3 sync s3://<source_bucket> <local_destination>

```shell
aws s3 sync s3://mybucket ./     # 将mybucket中的所有对象下载到当前目录
```
>aws s3 cp s3://BUCKETNAME/PATH/TO/FOLDER LocalFolderName --recursive

这将指示CLI在BUCKETNAME存储桶的PATH/TO/FOLDER目录中以递归方式下载所有文件和文件夹。  

## 三、python库boto3
　　参考:[AWS S3 - 使用python从S3下载数据](https://blog.csdn.net/xuezhangjun0121/article/details/119001173)   
### 3.1、文件下载
　　download_file方法
```python
import boto3

s3 = boto3.client('s3')
s3.download_file('BUCKET_NAME', 'OBJECT_NAME', 'FILE_NAME')
```
　　download_fileobj方法   
　　该方法必须以二进制形式打开。  
```python
s3 = boto3.client('s3')
with open('FILE_NAME', 'wb') as f:
    s3.download_fileobj('BUCKET_NAME', 'OBJECT_NAME', f)
```
