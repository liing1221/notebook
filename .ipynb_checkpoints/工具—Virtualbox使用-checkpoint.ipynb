## 一、Virtualbox安装设置
### 1、安装
　　官网下载Virtualbox后，直接点击默认安装即可！
### 2、设置
#### 2.1、虚拟机位置设置
　　自定义虚拟机存储位置，减少C盘资源占用。<br>
  　![虚拟机存储设置](./images/Virtualbox_设置虚拟机存储位置.jpg)
#### 2.2、虚拟机全局网络设置
　　nat网络(虚拟机访问互联网，使用10.0.2.x段)<br>
　　打开 VirtualBox   - 管理 - 全局设置 - 网络 ，没有NatNetwork的话,就手动添加一个，如下图配置：<br>
　　![虚拟机存储设置](./images/Virtualbox_全局网络设置.png)
#### 2.3、虚拟机虚拟网卡
　　打开 virtualBox  - 管理 -  主机网络管理器,点击创建(仅主机网络Host-only),点击手动配置网卡,启用DHCP。配置好网卡信息后，DHCP服务器勾选启用服务器，系统就会自动分配地址。<br>
　　![虚拟机存储设置](./images/Virtualbox_虚拟网卡创建设置.png)
#### 2.4、宿主机环境变量设置
　　Virtualbox 安装完成后，把安装路径（主目录）添加到系统环境变量Path,可在命令行使用VirtualBox的命令，如：后台运行虚拟机的命令VBoxManage等!。<br>

## 二、安装虚拟机Centos7
### 1、创建虚拟机
#### 1.1 创建设置虚拟机
　　根据主机系统，[下载Centos7](https://www.centos.org/download/)。<br>
  　点击Virtualbox的 创建 按钮，创建新的虚拟机。<br>
　　![虚拟机存储设置](./images/Virtualbox_创建虚拟机_设置名称存储.jpg)
　　补充：CentOS Linux 的发行版本是通过编译 Red Hat, Inc 公开提供的 Red Hat 企业级 Linux 开源代码而产生的。所以这里根据官方的说法应该选择redhat的发行版（其实这里选择Other Linux来创建虚拟机也没有什么问题）。<br>
　　点击： 下一步<br>
#### 1.2、设置虚拟机内存大小
　　按照主机内存，给虚拟机设置合适的内存大小（1—4G）。<br>
　　点击： 下一步<br>
#### 1.3、设置磁盘大小
　　按需要与主机硬盘大小设置合适的磁盘大小、存储位置。<br>
　　点击： 下一步<br>
#### 1.4、其他步骤
　　按默认点击：下一步，直至完成安装<br>
### 2、虚拟机设置
#### 2.1、选择安装镜像
　　选中新创建的虚拟机，点击右侧的设置，选择存储选项，在点击存储介质中的控制器栏目，点击栏目右侧的‘+’图标，添加注册虚拟机系统镜像（若已有镜像，直接选择即可）。<br>
　　![虚拟机存储设置](./images/Virtualbox_启动虚拟机_选择安装镜像.png) 
#### 2.2、虚拟机网络设置
　　选中新虚拟机，点击设置，选中网络选项，配置虚拟机的网络连接。<br>
##### 2.2.1、桥接网卡连接模式（单网卡）
  　![虚拟机存储设置](./images/Virtualbox_设置虚拟机网络.png)
##### 2.2.2、NAT网络模式（双网卡）
　　接下来对CentOS7虚拟机网络设置，添加网卡1和网卡2，分别选择连接方式和对应的网卡：<br>
　　有时候选择NAT网络，OK按钮显示为灰色不可用状态，那就选择 网路地址转换(NAT) ,有时候默认就是这个，我安装的时候默认就是这个。<br>
  　![虚拟机存储设置](./images/Virtualbox_NAT网络网卡1设置.png)
   ![虚拟机存储设置](./images/Virtualbox_NAT网络网卡2设置.png)
### 3、启动新的虚拟机
　　选中新创建的虚拟机，点击右侧的启动按钮。<br>
#### 3.1、启动虚拟机，选择需要的系统镜像安装。
#### 3.2、虚拟机设置
　　可设置项有：语言界面、用户、安装位置等。<br>
   ![虚拟机存储设置](./images/Virtualbox_虚拟机设置用户与密码.png)
#### 3.3、虚拟机设置系统安装位置
   ![虚拟机存储设置](./images/Virtualbox_虚拟机设置系统安装位置.png)
#### 3.4、虚拟机网卡与网络设置
　　进入虚拟机系统，打开终端，cd 到 /etc/sysconfig/network-scripts 文件夹下，使用vi 打开ifcfg-enp0s3(可查看实际本机的网卡名称)，设置虚拟机网卡网络。<br>
　　![root配置虚拟机网卡信息](./images/Virtualbox_虚拟机系统登录配置网络连接.png)<br>
　　![虚拟机网卡信息配置](./images/Virtualbox_虚拟机网卡配置.png)<br>
　　注意:是否已经将ONBOOT=no改为yes（最开始默认为NO，yes意思是，“BOOT开机”时候就能使用这张网卡），再添加BOOTPROTO为dhcp（只有“静态static”和“动态dhcp”两种方式），保存，退出。<br>
　　重启虚拟机网络，验证配置成功。<br>
  >service network restart　&　systemctl restart network<br>
  ip addr　　# 查看网卡信息是否与配置一致

![虚拟机网络验证](./images/Virtualbox_虚拟机网络验证.jpg)<br>
　　安装ifconfig(Centos7的知识)<br>
　　使用yum provides/whatprovides ifconfig 反查询，来查看那个包提供ifconfig，并安装该包。<br>
>yum whatprovides ifconfig<br>
yum install net-tools<br>
ifconfig<br>

### 4、虚拟机开机后台运行设置
　　VirtualBox 虚拟机开机后的显示界面，不仅不能够全屏而且占用桌面资源，且虚拟机home键（键盘右侧Ctrl键）使用起来，不是很方便，那么只需要VirtualBox帮我们在后台启动运行虚拟机，然后再通过xshell等ssh工具连接使用虚拟机不是很方便吗！<br>
#### 4.1 虚拟机无界面运行设置
　　命令行执行如下VirtualBox命令，设置无界面是用虚拟机：<br>
>VBoxManage.exe startvm \<uuid\>|\<name\> \[--type gui|sdl|headless\]<br>

![虚拟机无界面运行](./images/Virtualbox_后台运行虚拟机.jpg)<br>

#### 4.2 设置虚拟机开机自启动
　　上面虚拟机开机看不到窗口了，但是还是需要命令行手动输入启动。如果能开机自启动是不是就更好了。<br>
　　1、首选我们要做一个bat脚本：vboxStart.bat<br>
　　2、里面写入上面的命令：D:\VirtualBox\VBoxManage.exe startvm study --type headless ;若已注册环境变量，可直接写：VBoxManage startvm study --type headless<br>
　　3、把vboxStart.bat放到C:\ProgramData\Microsoft\Windows\Start Menu\Programs\Startup，这个目录下的脚本是开机的时候自动运行的哦.<br>
  例如，我的写法：<br>
>\#  路径及虚拟机名字问题自己改 <br>
D:<br>
cd D:\VirtualBox<br>
VBoxManage.exe startvm "Centos7" -type headless<br>

#### 4.3 headless模式的其他操作
　　对于开启了headless模式的虚拟机，我们可以用下列命令对其进行关闭重启等操作。<br>
>VBoxManage list vms　　　　#查看已创建的虚拟机列表
VBoxManage controlvm \<uuid\>|\<name\>     pause|resume|reset|poweroff|savestate|<br>
vboxmanage controlvm \<name\> poweroff　　　# 例如:关闭headless虚拟机<br>
vboxmanage sharedfolder add centos --name *share* --hostpath *~/share*　　　# 想使用共享目录<br>
mount -t vboxsf share /mnt/　　　　# 登录成功后挂载共享目录<br>
