# 日志查看命令
　　　在程序的开发、测试、部署、运维等过程中，经常需要实时查看日志文件的内容。下面我们来介绍一些好用、高效的文件处理命令。


## 一、tail命令
　　tail命令是实时显示日志文件的最常用解决方案。但是，显示该文件的命令有两个版本，如下面的示例所示。
　　在第一个示例中，命令tail需要-f参数来跟踪文件的内容。  
```shell
$ sudo tail -f /var/log/apache2/access.log     # 实时监控Apache日志
```
　　第二个版本实际上是一个命令本身：tailf。  
　　你不需要使用该-f开关，因为该命令是与-f参数一起内置的。  
```shell
$ sudo tailf /var/log/apache2/access.log
```
　　实时Apache日志监控  
　　通常，logrotate实用程序会在Linux服务器上频繁地轮换日志文件。要查看每日基础上轮换的日志文件，可以使用tail -F命令。  
　　tail -F将跟踪正在创建的新日志文件，并开始跟踪新文件而不是旧文件。   
```shell
$ sudo tail -F /var/log/apache2/access.log
```
　　但是，默认情况下，tail命令将显示文件的最后10行。例如，如果你只想实时查看日志文件的最后两行，请使用-n与该-f标志结合的文件，如下例所示。  
```shell
$ sudo tail -n2 -f /var/log/apache2/access.log     # 查看最后两行日志
```

## 二、multitail命令
　　multitail命令用于实时监控多个日志文件。该命令的名称意味着multitail实用程序可以实时监视和跟踪多个文件。Multitail还允许你在受监视的文件中来回导航。   
　　要在基于Debian和RedHat的系统中安装mulitail实用程序，请发出以下命令。   
```shell
$ sudo apt install multitail [On Debian＆Ubuntu]
$ sudo yum install multitail [On RedHat＆CentOS]
$ sudo dnf install multitail [On Fedora 22+ version]
```
　　要同时显示两个日志文件的输出，请执行以下示例中所示的命令。
```shell
$ sudo multitail /var/log/apache2/access.log /var/log/apache2/error.log　　　　　# 多点监控日志
```

## 三、lnav命令
　　lnav用于实时监控多个日志文件，类似于multitail命令。Lnav实用程序还可以观看和跟踪多个文件并实时显示其内容。
通过发出以下命令在基于Debian和RedHat的Linux发行版中安装lnav实用程序。
```shell
$ sudo apt install lnav [On Debian＆Ubuntu]
$ sudo yum install lnav [On RedHat＆CentOS]
$ sudo dnf install lnav [On Fedora 22+ version]
```
　　通过发出命令同时观察两个日志文件的内容，如下例所示。
```shell
$ sudo lnav /var/log/apache2/access.log /var/log/apache2/error.log　　　　# lnav - 实时日志监控
```
## 四、less命令 
　　less显示日志文件的实时输出
```shell
$ sudo less + F /var/log/apache2/access.log      # 使用less命令监视日志
```
