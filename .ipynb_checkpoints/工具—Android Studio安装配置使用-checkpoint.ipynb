## 一、概览
　　Android Studio 是开发 Android 应用程序的官方 IDE，基于 Intellij IDEA。你可以从官网Android Studio下载下载最新版本的 Android Studio。如果你是在 Windows 上安装 Android Studio 的新人，找到名为 android-studio-bundle-135.17407740-windows.exe 文件下载，并通过 Android Studio 向导指南运行。<br>
　　如果要在 Mac 或者 Linux 上安装 Android Studio，可以从Android Studio Mac 下载或者Android Studio Linux 下载来下载最新版本，查看随下载文件提供的说明。除了 IntelliJ 强大的代码编辑器和开发者工具，Android Studio 还提供更多可提高 Android 应用构建效率的功能：<br>
　　1)、基于 Gradle 的灵活构建系统<br>
　　2)、快速且功能丰富的模拟器<br>
　　3)、统一的环境（供您开发适用于所有 Android 设备的应用）<br>
　　4)、Apply Changes 功能可将代码和资源更改推送到正在运行的应用，而无需重启应用<br>
　　5)、代码模板和 GitHub 集成，可协助您打造常见的应用功能及导入示例代码<br>
　　6)、大量的测试工具和框架<br>
　　7)、Lint 工具，能够找出性能、易用性和版本兼容性等方面的问题<br>
　　8)、C++ 和 NDK 支持<br>
　　9)、内置对 Google Cloud Platform 的支持，可轻松集成 Google Cloud Messaging 和 App Engine<br>
　　[阅读详情](https://developer.android.google.cn/studio/intro)<br>
## 二、安装Android Studio
　　确保在安装 Android Studio 之前，已经安装好 Java JDK，并配置好Java开发环境。<br>
　　直接电脑桌面windows+R 输入cmd：  上代码：分别输入java 、javac 、javadoc、java -version  + enter键   成功的有输出，说明java环境安装成功。<br>
### 2.1、Windows 系统安装 Android Studio：
　　下载Android Studio安装包。<br>
　　如果您下载了 .exe 文件（推荐），请双击以启动该文件。<br>
　　如果您下载了 .zip 文件，请解压缩 ZIP 文件，将 android-studio 文件夹复制到 Program Files 文件夹中，然后依次打开 android-studio > bin 文件夹并启动 studio64.exe（对于 64 位计算机）或 studio.exe（对于 32 位计算机）。<br>
　　按照 Android Studio 中设置向导的指示安装它推荐的所有 SDK 软件包。<br>
　　建议自定义安装包的安装路径， 如：D:\android\android studio。  
　　建议自定义android SDK的安装路径， 如：:\Android\Sdk。  
　　IDE内存选择：最好不要低于2G（2048MB），可根据自己的系统内存大小适当增大。   
### 2.2、安装gradle
　　第一次使用Studio,IDE会自动安装gradle.  
　　安装失败：Gradle project sync failed!...  
　　解决办法：1、确认自己的gradle版本。在gradle-wrapper.properties文件查看。  
　　　　　　　2、[下载gradle的对应安装包](https://services.gradle.org/distributions/)，并复制到Studio的默认下载路径(清空该路径下其他非安装包文件)。
　　　　　　　3、重启Android Studio，等待安装包安装完成。   
　　修改默认下载路径：在 project 的 build.gradle中修改。  
'''gradle  
allprojects {  
    repositories {  
        //maven{ url 'http://maven.oschina.net/content/groups/public/'}  
        maven { url 'https://maven.aliyun.com/repository/public/' }  
        //jcenter()  
        google()  
    }  
}   
'''  

## 二、为Android Studio安装模拟器

　　可参考：[Android Studio 模拟器的选择和安装](https://blog.csdn.net/qq_33505204/article/details/78452286)  
打开Tools -> SDK Manager, 勾选并安装需要的安卓版本，点击apply。

### 2.1、模拟器的选择
1、Intel CPU主机，需要BIOS中开启处理器虚拟化技术。  
2、非Intel CPU主机，可选择arm模拟器或第三方模拟器Genymotion(依赖VirtualBox添加对应手机的Virtual device, 以加载手机虚拟机)。  
3、安装真机模拟器

### 2.2、安装模拟器
1、打开android studio，在tools选项卡选择AVD Manager，进入模拟器管理界面。
2、在弹窗中选择Create Virtual Device添加新模拟器设备
3、选择设备：在新弹窗中选择想要的设备，然后点击 next > next > Finish 即可。
4、异常处理1：Error while waiting for device: The emulator process for AVD Pixel_2_API_28 was killed.
异常原因：
    1.Android Emulater没有安装。
    2.HAXM需要更新或重新安装

下面看看那个没安装：File > Settings > System Settings > Android SDK > SDK Tools > 没安装的安装，版本过低的升级版本 > 选中红框内的内容 > Apply。
5、异常处理2：Unable to install Intel® HAXM VT-x is disabled in BIOS.
异常原因：未开启处理器虚拟化技术
解决：重启计算机，进入BIOS，在系统设置中找到处理器虚拟化技术并开启，F10保存退出。

6、异常处理3：Unable to determine application id: com.android.tools.idea.run.ApkProvisionException: The currently selected variant "debug" uses split APKs, but none of the 1 split apks are compatible with the current device with ABIs 「"x86"」.
异常原因：项目未兼容x86  

'''
//  build.gradle>android>defaultConfig>ndk>新增'x86'，如下：  
android {  
    compileSdkVersion 30//项目的编译版本  
    defaultConfig {  
        applicationId "com.scc.demo"//包名  
        minSdkVersion 23//最低的兼容的Android系统版本  
        targetSdkVersion 30//目标版本，表示你在该Android系统版本已经做过充分的测试  
        versionCode 1//版本号  
        versionName "1.0.0"//版本名称  
        ndk {  
            abiFilters 'armeabi-v7a', 'arm64-v8a','x86'  
            //运行环境，要上传Google Play必须兼容64位，这里仅兼容ARM架构  
            //对于ARM架构，32 位库位于armeabi-v7a 中。64 位等效项是arm64-v8a。  
            //对于x86体系结构，查找x86（用于 32 位）和 x86_64（用于 64 位）。  
        }  
    }  
}  
'''
