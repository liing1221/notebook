## Linux—tar命令详解
　　参考：[Linux中常用的tar解压打包命令语法介绍](https://blog.csdn.net/guo_qiangqiang/article/details/87934655)<br>
　　　　　[tar.xz文件如何解压](https://blog.csdn.net/rheostat/article/details/7614451)<br>

### 一、简介
　　linux下最常用的打包程序就是tar了，使用tar程序打出来的包我们常称为tar包，tar包文件的命令通常都是以.tar结尾的。生成tar包后，就可以用其它的程序来进行压缩。<br>

### 二、命令格式
>tar [必要参数] [选择参数] [文件] <br>

### 三、命令功能
　　用来压缩和解压文件。tar本身不具有压缩功能。他是调用压缩功能实现的 <br>

### 四、命令参数
　　必要参数有如下：
>-A 新增压缩文件到已存在的压缩<br>
-B 设置区块大小<br>
-c 建立新的压缩文件<br>
-d 记录文件的差别<br>
-r 添加文件到已经压缩的文件<br>
-u 添加改变了和现有的文件到已经存在的压缩文件<br>
-x 从压缩的文件中提取文件<br>
-t 显示压缩文件的内容<br>
-z 支持gzip解压文件<br>
-j 支持bzip2解压文件<br>
-Z 支持compress解压文件<br>
-v 显示操作过程<br>
-l 文件系统边界设置<br>
-k 保留原有文件不覆盖<br>
-m 保留文件不被覆盖<br>
-W 确认压缩文件的正确性<br>

　　可选参数如下：
>-b 设置区块数目<br>
-C 切换到指定目录<br>
-f 指定压缩文件<br>
--help 显示帮助信息<br>
--version 显示版本信息<br>

### 五、常见解压/压缩命令

#### 1、tar 
解包：tar xvf FileName.tar<br>
打包：tar cvf FileName.tar DirName<br>
<br>
（注：tar是打包，不是压缩！）<br>
#### 2、.gz
解压1：gunzip FileName.gz<br>
解压2：gzip -d FileName.gz<br>
压缩：gzip FileName<br>
#### 3、.tar.gz 和 .tgz
解压：tar zxvf FileName.tar.gz<br>
压缩：tar zcvf FileName.tar.gz DirName<br>
#### 4、.bz2
解压1：bzip2 -d FileName.bz2<br>
解压2：bunzip2 FileName.bz2<br>
压缩： bzip2 -z FileName<br>
#### 5、.tar.bz2
解压：tar jxvf FileName.tar.bz2<br>
压缩：tar jcvf FileName.tar.bz2 DirName<br>
#### 6、.bz
解压1：bzip2 -d FileName.bz<br>
解压2：bunzip2 FileName.bz<br>
压缩：未知<br>
#### 7、.tar.bz
解压：tar jxvf FileName.tar.bz<br>
压缩：未知<br>
#### 8、.Z
解压：uncompress FileName.Z<br>
压缩：compress FileName<br>
#### 9、.tar.Z
解压：tar Zxvf FileName.tar.Z<br>
压缩：tar Zcvf FileName.tar.Z DirName<br>
#### 10、.zip
解压：unzip FileName.zip<br>
压缩：zip FileName.zip DirName<br>
#### 11、.rar
解压：rar x FileName.rar<br>
压缩：rar a FileName.rar DirName <br>

### 六、使用实例

#### 1、将文件全部打包成tar包
>tar -cvf log.tar log2012.log    仅打包，不压缩 <br>
tar -zcvf log.tar.gz log2012.log   打包后，以 gzip 压缩 <br>
tar -zcvf log.tar.bz2 log2012.log  　　打包后，以 bzip2 压缩 <br>

　　在参数 f 之后的文件档名是自己取的，我们习惯上都用 .tar 来作为辨识。 如果加 z 参数，则以 .tar.gz 或 .tgz 来代表 gzip 压缩过的 tar包； 如果加 j 参数，则以 .tar.bz2 来作为tar包名。<br>

#### 2、查阅上述 tar包内有哪些文件
>tar -ztvf log.tar.gz<br>

　　由于我们使用 gzip 压缩的log.tar.gz，所以要查阅log.tar.gz包内的文件时，就得要加上 z 这个参数了。<br>

#### 3、将tar 包解压缩
>tar -zxvf /opt/soft/test/log.tar.gz<br>
 
　　在预设的情况下，我们可以将压缩档在任何地方解开的<br>

#### 4、只将 /tar 内的 部分文件解压出来
>tar -zxvf /opt/soft/test/log.tar.gz log2012.log<br>

　　我可以透过 tar -ztvf 来查阅 tar 包内的文件名称，如果单只要一个文件，就可以透过这个方式来解压部分文件！<br>

#### 5、文件备份下来，并且保存其权限
>tar -zcvpf log31.tar.gz log2014.log log2015.log log2016.log <br>

　　这个 -p 的属性是很重要的，尤其是当您要保留原本文件的属性时<br>

#### 6、在 文件夹当中，比某个日期新的文件才备份
>tar -N "2012/11/13" -zcvf log17.tar.gz test<br>

#### 7、备份文件夹内容是排除部分文件
>tar --exclude scf/service -zcvf scf.tar.gz scf/*<br>

#### 8、将文件解压到指定目录下
>tar zxvf test.tgz -C 指定目录<br>
eg：tar zxvf /usr/local/jdk-8u161-linux-x64.tar.gz -C /usr/local/jdk1.8 <br> 

　　将压缩文件jdk-8u161-linux-x64.tar.gz解压到/usr/local/jdk1.8目录下<br>
