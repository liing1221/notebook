# Pip 安装工具的使用
## 一、pip下载安装
### 1.1 pip下载
　　进入 https://pypi.python.org/pypi/pip ，下载 pip-xxx.tar.gz 压缩包文件<br>
### 1.2  Linux安装pip
>tar -xzvf pip-1.5.4.tar.gz　　　# 解压<br>
cd pip-1.5.4　　　# 进入解压文件<br>
python setup.py install　　　　# 安装<br>

### 1.3 升级pip
>python -m pip install --upgrade pip<br>

### 1.4、修改pip的默认安装路径
　　windows下，pip默认安装在C盘的用户主目录的appdata目录下，占用C盘的空间。此时可以修改pip的安装路径，把python库安装到其他空间充足的目录中。  
```shell
$ python -m site       # 查看安装路径设置

sys.path = [
    'D:\\Apps\\Miniconda3\\python39.zip',
    'D:\\Apps\\Miniconda3\\DLLs',
    'D:\\Apps\\Miniconda3\\lib',
    'D:\\Apps\\Miniconda3',
    'D:\\Apps\\Miniconda3\\Scripts',
    'D:\\Apps\\Miniconda3\\lib\\site-packages',
    'D:\\Apps\\Miniconda3\\lib\\site-packages\\win32',
    'D:\\Apps\\Miniconda3\\lib\\site-packages\\win32\\lib',
    'D:\\Apps\\Miniconda3\\lib\\site-packages\\Pythonwin',
]
USER_BASE: 'D:\\Apps\\Miniconda3\\Lib\\site-packages' (exists)
USER_SITE: 'D:\\Apps\\Miniconda3\\Scripts' (exists)
ENABLE_USER_SITE: True
# 其中USER_BASE和USER_SITE是pip的默认安装路径

$ python -m site -help   # 查看配置 USER_BASE 与 USER_SITE 的文件路径

D:\Apps\Miniconda3\lib\site.py [--user-base] [--user-site]    # 此处，site.py 文件即为要修改配置的文件

Without arguments print some useful information
With arguments print the value of USER_BASE and/or USER_SITE separated
by ';'.

Exit codes with --user-base or --user-site:
  0 - user site directory is enabled
  1 - user site directory is disabled by user
  2 - user site directory is disabled by super user
      or for security reasons
 >2 - unknown error
```
　　找到，site.py 文件，打开，并找到 USER_BASE = None、USER_SITE = None的两行修改。   
　　注意：此处坑两点，一是 user_base 需设置为python安装目录下的Scripts文件夹得路径，否则后续使用安装的库时会有无法导包的问题；二是，由于配置路径不再是C盘的用户主目录，存在pip安装没有写权限的问题，此时需要给相关的文件（python安装目录下的etc、share、以及配置路径三处）路径赋予写权限。具体操作是：文件夹上右键 -> 属性 -> 安全 -> 编辑 -> 组与用户名中选择Users后在其权限中允许修改、写入 -> 应用 -> 确定 即可    
　　此时，再通过 python -m site 查看，可发现配置已生效，赶紧 pip install 个包试一下吧！  

## 二、pip使用详解
### 2.1 pip安装包
>pip install xxx　　　# 安装包名<br>
>　\[...\]<br>
　Successfully installed SomePackage    # 安装成功<br>

>pip show --files 安装包名

### 2.3 pip检查哪些包需要更新
>pip list --outdated

### 2.4 pip升级包
>pip install --upgrade 要升级的包名

### 2.5 pip卸载包

### 2.7 search功能补充
　　根据 PyPi 官方文档（20210326更新），由于持续的不明XMLRPC请求滥用，Python.org的XMLRPC将永久停止响应。也就意味着 PyPi 关闭了 pip 的 search 功能。 继续使用 pip search 将报错 ERROR: XMLRPC request failed [code: -32500]。<br>
　　要想使用 pip search 的功能，就需要使用 pip_search 替代了！<br>
　　安装：<br>
> pip install pip_search

　　使用：<br>
> pip_search requests

修改配置默认源：<br>
1、需要创建或修改配置文件（一般都是创建）<br>
linux的文件在~/.pip/pip.conf<br>
windows在%HOMEPATH%\pip\pip.ini）<br>
2、修改内容为：<br>
\[global\]<br>
index-url = http://pypi.douban.com/simple<br>
\[install\]<br>
trusted-host=pypi.douban.com<br> 
这样在使用pip来安装时，会默认调用该镜像。<br>
>#!/usr/bin/python <br>
import os<br> 
package = raw_input("Please input the package which you want to install!\n")<br>
command = "pip install %s -i http://pypi.mirrors.ustc.edu.cn/simple --trusted-host pypi.mirrors.ustc.edu.cn" % package<br>
os.system(command)<br>

## pip 工具使用

### pip更新升级
>pip -V　　　# pip版本查看
>pip install --upgrade pip

### pip安装错误处理<br>
1、错误：Retrying (Retry(total=4, connect=None, read=None, redirect=None, status=None)) after connection broken by 'ConnectTimeoutError<br>
解决办法：pip install selectivesearch -i http://pypi.douban.com/simple --trusted-host pypi.douban.com<br>
2、错误：信任与安全主机错误<br>
解决办法： pip install web.py -i http://pypi.douban.com/simple --trusted-host pypi.douban.com<br>
3、错误：Fatal Python error: init_sys_streams: can't initialize sys standard streams；LookupError: unknow encoding: 65001<br>
解决办法：命令行执行set pythonioencoding=utf-8 <br>
4、错误：Python--Cmd窗口运行Python时（或者PIP安装第三方库时）提示：Fatal Python error: Py_Initialize: can't initialize sys standard streams LookupError: unknown encoding: cp65001<br>
参考：[Cmd窗口运行Python时：unknown encoding: cp65001](https://www.bbsmax.com/A/Vx5M013gJN/)<br>
解决办法：（结合chcp查看cmd代码页，并修改为合适编码）set pythonioencoding=utf-8 <br>
5、错误：python setup.py egg_info ...”等等，但是倒数几行显示：“gbk” codec con't...（如安装 fitter 包）  
解决办法：下载fitter 压缩包，解压打开，找到setup.py文件打开，修改文件内open()方法的打开编码方式，设置为encoding='utf-8', 保存退出。然后再使用 "pip install 文件路径+包名称" 的方式安装第三方包。   
```shell
pip install d:\adnconda\fitter-1.3.0
```
