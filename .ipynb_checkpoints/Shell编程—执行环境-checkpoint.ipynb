# Shell执行环境

　　Shell是一个命令行解释器，它接收应用程序/用户命令，然后调用操作系统内核。
　　Shell还是一个功能相当强大的编程语言，易编写，易调试，灵活性强。
Shell是一个用 C 语言编写的程序，它是用户使用 Linux 的桥梁。Shell 既是一种命令语言，又是一种程序设计语言。
Shell 是指一种应用程序，这个应用程序提供了一个界面，用户通过这个界面访问操作系统内核的服务。  
　　Shell脚本文件就是一个文本文件, 后缀名建议使用 .sh 结尾。   
## 1、Shell解释器
　　Shell 程序的执行依赖shell解释器，常用的是 bash 。查看系统的shell解释器，以及默认shell解释器：  

```shell
liing@ubuntu: cat /etc/shells 
# /etc/shells: valid login shells
/bin/sh
/bin/bash
/bin/rbash
/bin/dash
liing@ubuntu:~$ echo $SHELL
/bin/bash
```

## 2、Shell脚本执行
　　Shell文件的常用的3种执行方式  
1. bash scriptName 或 sh scriptName，当脚本文件本身没有执行权限时常使用的方法
2. path/scriptName 或 ./scriptName，在当前路径下执行脚本，需要将脚本文件的权限改为可执行
3. source scriptName 或 . scriptName，通常是使用source或 "." 读入或加载指定的Shell脚本文件，然后依次执行指定的Shell脚本中的语句。source是一个内建shell。

　　注意：为执行脚本添加执行权限。  
>chmod +x xxx.sh    # xxx.sh文件添加可执行权限   
chmod a+x xxx.sh    # xxx.sh文件当前用户添加可执行权限   
chmod 777 xxx.sh   # 最高权限   

## 3、Shell脚本规范
    一个规范的Shell脚本会在**第一行**指出由哪个程序（解释器）来执行脚本中的内容,从第二行开始编写脚本内容。如，使用Linux bash的编程一般为：#！/bin/bash 。此外，还有 #! /bin/sh ，bash与sh的区别是：sh为bash的软连接。
    　　编写规范：  
```shell
#!/bin/bash				[指定告知系统当前这个脚本要使用的shell解释器]
```
    文件命名规范：
Shell相关指令
文件名.sh     .sh是linux下bash shell 的默认后缀

## 4、Shell 脚本注释
　　shell脚本的注释分两种：单行注释和多行注释。　　
### 4.1、单行注释
　　单行注释是在注释语句行首位置添加‘#’字符；如果想在代码行后面添加注释，则在键入‘#’前，先键入空格。  

### 4.2、多行注释
　　语法如下：  
```shell
:<<!
注释语句块
!
```
## 5、Shell 配置文件
　　一般情况下，我们都是直接针对全局配置进行操作。  
### 5.1、全局配置文件
```shell
/etc/profile
/etc/profile.d/\*.sh
/etc/bashrc
```
### 5.2、用户配置文件
```shell
当前用户/.bash_profile
当前用户/.bashrc
```
