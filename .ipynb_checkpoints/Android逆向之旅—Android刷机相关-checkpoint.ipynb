## 三、工具使用
### 3.1、Fastboot 
　　参考:[手机刷机调试命令：fastboot详细说明](https://www.toutiao.com/article/7055460122310574596/?group_id=7055460122310574596&wid=1658236374435)  
```shell
fastboot reboot                 # 重新启动到正常模式
fastboot reboot bootloader      # 重新启动到引导模式
fastboot reboot recovery        # 重新启动到恢复模式

fastboot flashing lock          # 给BL上锁
fastboot flashing unlock        # 解BL锁

# android 6.0以前
fastboot flashing lock          # 给BL上锁
fastboot flashing unlock        # 解BL锁
```

## 四、刷机方法
### 4.1、进入开发者模式
　　在手机，打开设置 -> 关于手机 -> 多次点击 版本号 -> 直到提示进入开发者模式。  
### 4.2、刷机准备：打开USB调试
### 4.3、刷机准备：Bootloader(OEM)解锁 
1. 从你的设备中删除Google帐户和任何类型的屏幕锁定（指纹，PIN码，图案等）。
2. 从设备中弹出SIM卡。
3. 重置设备。在设置向导中，跳过所有内容，不要连接到WiFi，不要添加指纹或任何类型的屏幕锁定。
4. 转到开发人员选项并启用USB调试。
5. 将手机连接到PC。
6. 在adb目录中打开CMD并键入adb shell pm uninstall --user 0 com.android.phone
7. 重新启动设备。
8. 连接到WiFi，打开Chrome，然后转到google.com（或任何网站）。
9. 转到开发人员选项并启用OEM解锁。
10. 重新启动进入引导程序并通过CMD运行fastboot oem unlock或是fastboot flashing unlock
　　注意：解锁BL会重置手机

　　OEM解锁：在手机，打开设置 -> 系统 -> 高级 -> 开发者选项 ->找到OEM解锁:打开。  
　　BL解锁：  
```shell
      adb devices     # 检查手机是否已连接
      adb reboot bootloader    # 先进入 fastboot mode
      fastboot flashing unlock     # 解BL锁
```
　　注：若以上命令卡在 < waiting for any device > 状态，则使用 fastboot devices 检查设备是否存在，若不存在，则可能是未安装USB驱动(linux主机未安装驱动也会出现该问题)     
　　USB 驱动下载：https://developer.android.google.cn/studio/run/win-usb     
　　USB 驱动安装：PC进入 设备管理器 -> 其它设备下 -> 查看相应的设备，本人设备为 Pixle3 XL -> 右键更新驱动程序 -> 选择浏览我的电脑以查找驱动程序 -> 将下载驱动解压缩路径选中 -> 点击下一步 -> 安装即可。   
### 4.4、fastbool模式启动：
　　手机端方法一:　手机关机　-> 　电源 + 音量减 -> 启动到fastboot模式   
　　PC端方法二:　adb reboot bootloader    
### 4.5、刷入镜像包
　　1. 解压缩 Google 官方镜像.zip 后，进入解压缩后的目录: Windows 下CMD中执行 flash-all.bat; Linux 下执行 flash-all.sh, 刷入下载的官方镜像，等待刷入即可。   
　　2. 解压缩 Google 官方镜像.zip 后，里面还有个zip文件，继续解压缩后，进入解压缩后的目录，执行指令将文件 boot.img 发送到手机 /sdcard/download/ 路径下。    
>adb push boot.img /sdcard/download/   

### 4.6、Magisk 解锁文件
　　1、手机安装 magisk-23.apk  
　　2、打开 Magisk, 在Magisk栏点击: 安装 -> 选择并修补一个文件 -> 选择 /sdcard/download/boot.img -> 点击开始,即可。  
　　3、上一步在原位置生成一个解锁后的 img 文件：magisk_patched-xxx_xxx.img, 下载到电脑上(注意与 fastboot 工具在同一目录)：  
>adb pull /sdcard/download/magisk_patched-xxx_xxx.img ./  

### 4.7、刷入解锁包
```shell
 adb reboot bootloader           # 先进入 fastboot mode
 fastboot flash boot magisk_patched_xxx_xxx.img     # 将解锁包刷入系统的boot分区，使刷机长期有效。
 
 fastboot flash boot magisk_patched-xxx_xxx.img      # 执行命令刷入解锁包  
 fastboot flash recovery recovery.img      # 将当前目录下的recovery.img文件刷进系统的recovery分区。  
```
### 4.8、查看验证
　　手机开机，打开 magisk 应用，“当前”状态显示 magisk 版本号，表示成功。  
　　获取root权限: adb shell -> 进入android命令行,执行：su -> 获取root权限 -> magisk 显示\#号(已root)。
