# dateutil库详解
　　　参考：[Python时间处理-dateutil模块
](https://www.jianshu.com/p/f29dddce3a9a)<br>

## 一、简介
　　dateutil模块主要有两个函数，parser和rrule。<br>
　　其中parser是根据字符串解析成datetime，而rrule则是根据定义的规则来生成datetime。<br>
## 二、安装
　　直接用easy_install或pip在线安装<br>
>easy_install python-dateutil<br>
pip install python-dateutil<br>

## 三、dateutil.parser
　　parser是根据字符串解析成datetime,字符串可以很随意，可以用时间日期的英文单词，可以用横线、逗号、空格等做分隔符。<br>
　　没指定时间默认是0点，没指定日期默认是今天，没指定年份默认是今年。<br>
> from dateutil.parser import parse<br>
parse("2018-10-21")　　　　# datetime.datetime(2018, 10, 21, 0, 0)<br>
parse("20181021")　　　　# datetime.datetime(2018, 10, 21, 0, 0)<br>
parse("2018/10/21")　　　　# datetime.datetime(2018, 10, 21, 0, 0)<br>
parse("10-21")　　　　# datetime.datetime(2018, 10, 21, 0, 0)<br>
parse("10/21")　　　　# datetime.datetime(2018, 10, 21, 0, 0)<br>

　　只有月日时，parser会将分隔符前面的数字解析为月份，后面的为日。<br>
　　当有年份时，在前面的月份超出范围时，会自动判断哪个是月哪个是日。<br>
> parse("3/8")　　　# datetime.datetime(2018, 3, 8, 0, 0)<br>
parse("8/3")　　　# datetime.datetime(2018, 8, 3, 0, 0)<br>
parse("3-8")　　　# datetime.datetime(2018, 3, 8, 0, 0)<br>
parse("8-3")　　　# datetime.datetime(2018, 8, 3, 0, 0)<br>
parse("21/10")　　　# ValueError: month must be in 1..12<br>
parse("3/8/2018")　　　# datetime.datetime(2018, 3, 8, 0, 0)<br>
parse("8/3/2018")　　　# datetime.datetime(2018, 8, 3, 0, 0)<br>
parse("2018/3/8")　　　# datetime.datetime(2018, 3, 8, 0, 0)<br>
parse("3/15/2018")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>

　　当前面的月份超过12时，parser会自动识别月和日<br>
> parse("15/3/2018")<br>
datetime.datetime(2018, 3, 15, 0, 0)<br>

　　但是当年份放在前面时，只能按年-月-日的顺序<br>
> parse("2018/15/3")<br>
ValueError: month must be in 1..12<br>

　　当分隔符为逗号时，只有月-日时，要把月放在前面；有年份时，年份要放在后面<br>
> parse("3,15")
datetime.datetime(2018, 3, 15, 0, 0)

　　只识别到了前面的日，月将本月11月作为默认月<br>
> parse("15,3")　　　# datetime.datetime(2018, 11, 15, 0, 0)<br>
parse("3,15,2018")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>
parse("2018,3,15")　　　# ValueError: ('Unknown string format:', '2018,3,15')<br>

　　parser还可以识别英文的月、日
> parse("Mar 15")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>

　　没有空格也可以
> parse("Mar15")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>
parse("15 Mar")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>
parse("Mar 1st")　　　# datetime.datetime(2018, 3, 1, 0, 0)<br>
parse("Mar 15 2018")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>
parse("2018 Mar15")　　　# datetime.datetime(2018, 3, 15, 0, 0)<br>

## 四、dateutil.rrule
　　rrule(self, freq, dtstart=None, interval=1, wkst=None,count=None, until=None, bysetpos=None,bymonth=None, bymonthday=None, byyearday=None, byeaster=None,byweekno=None, byweekday=None, byhour=None, byminute=None, bysecond=None,cache=False)<br>

|参数|说明|
|:---|:---|
|freq|可以理解为单位。可以是 YEARLY, MONTHLY, WEEKLY,DAILY, HOURLY, MINUTELY, SECONDLY。即年月日周时分秒|
|dtstart,until|是开始和结束时间|
|wkst|周开始时间|
|interval|间隔|
|count|指定生成多少个|
|byxxx|指定匹配的周期。比如byweekday=(MO,TU)则只有周一周二的匹配。byweekday可以指定MO,TU,WE,TH,FR,SA,SU。即周一到周日。|

> from dateutil import rrule<br>
\# 生成一个连续的日期列表<br>
list(rrule.rrule(rrule.DAILY,dtstart=parse('2018-11-1'),until=parse('2018-11-5')))<br>
[datetime.datetime(2018, 11, 1, 0, 0), <br>
datetime.datetime(2018, 11, 2, 0, 0), <br>
datetime.datetime(2018, 11, 3, 0, 0), <br>
datetime.datetime(2018, 11, 4, 0, 0), <br>
datetime.datetime(2018, 11, 5, 0, 0)]<br>
\# 间隔一天<br>
list(rrule.rrule(rrule.DAILY,interval=2,dtstart=parse('2018-11-1'),until=parse('2018-11-5')))<br>
[datetime.datetime(2018, 11, 1, 0, 0), <br>
datetime.datetime(2018, 11, 3, 0, 0), <br>
datetime.datetime(2018, 11, 5, 0, 0)]<br>
\# 只保留前3个元素<br>
list(rrule.rrule(rrule.DAILY,count=3,dtstart=parse('2018-11-1'),until=parse('2018-11-5')))<br>
[datetime.datetime(2018, 11, 1, 0, 0), <br>
datetime.datetime(2018, 11, 2, 0, 0), <br>
datetime.datetime(2018, 11, 3, 0, 0)]<br>
\# 只要周一的<br>
list(rrule.rrule(rrule.DAILY,byweekday=rrule.MO,dtstart=parse('2018-11-1'),until=parse('2018-11-5')))<br>
[datetime.datetime(2018, 11, 5, 0, 0)]<br>
\# 只要周六和周日的<br>
list(rrule.rrule(rrule.DAILY,byweekday=(rrule.SA,rrule.SU),dtstart=parse('2018-11-1'),until=parse('2018-11-5')))<br>
[datetime.datetime(2018, 11, 3, 0, 0), <br>
datetime.datetime(2018, 11, 4, 0, 0)]<br>
\# 以月为间隔<br>
list(rrule.rrule(rrule.MONTHLY,dtstart=parse('2018-3-15'),until=parse('2018-7-30')))<br>
[datetime.datetime(2018, 3, 15, 0, 0), <br>
datetime.datetime(2018, 4, 15, 0, 0), <br>
datetime.datetime(2018, 5, 15, 0, 0), <br>
datetime.datetime(2018, 6, 15, 0, 0), <br>
datetime.datetime(2018, 7, 15, 0, 0)]<br>

　　**计算时间差**<br>
　　rrule可计算出两个datetime对象间相差的年月日等时间数量。<br>
>\# 两个日期相差10天<br>
rrule.rrule(rrule.DAILY,dtstart=parse('2018-11-1'),until=parse('2018-11-10')).count()　　　　# 10<br>
\# 某个日期到今天相差多少天<br>
rrule.rrule(rrule.DAILY,dtstart=parse('2018-11-1'),until=datetime.date.today()).count()　　　　# 10<br>

　　**两个日期相差几个月**<br>
　　前一个月为m月，后一个月为n月，当日期不满整月时，差的月数按n-m算，当日期满整月后，差的月数按n-m+1算。<br>
　　差的年数同月数的情况一样。<br>
> rrule.rrule(rrule.MONTHLY,dtstart=parse('2018-3-15'),until=parse('2018-11-10')).count()　　　　# 8<br>
rrule.rrule(rrule.MONTHLY,dtstart=parse('2018-3-15'),until=parse('2018-11-20')).count()　　　　# 9<br>
rrule.rrule(rrule.YEARLY,dtstart=parse('2016-3-15'),until=parse('2018-2-10')).count()　　　　# 2<br>
rrule.rrule(rrule.YEARLY,dtstart=parse('2016-3-15'),until=parse('2018-3-15')).count()　　　　# 3<br>
