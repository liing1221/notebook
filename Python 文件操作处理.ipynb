#### 文件系统和文件<br>
　　在磁盘上读写文件的功能都是由操作系统提供的，现代操作系统不允许普通的程序直接操作磁盘，所以，读写文件就是请求操作系统打开一个文件对象（通常称为文件描述符），然后，通过操作系统提供的接口从这个文件对象中读取数据（读文件），或者把数据写入这个文件对象（写文件）。<br>
　　文件是计算机中由OS管理的具有名字的存储区域，在Linux中，文件是被看做字节序列.<br>
#### open() 函数<br>
open() :python内置函数，用于打开一个文件，创建一个 file 对象，相关的方法才可以调用它进行读写。<br>
完整语法格式为：<br>
>open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)<br>

参数说明：<br>
>file: 　　必需，文件路径（相对或者绝对路径）,一个包含了你要访问的文件名称的字符串值。<br>
mode: 　　可选，文件打开模式：只读，写入，追加等，默认文件访问模式为只读(r)。<br>
buffering:　　 设置缓冲；如果 buffering 的值被设为 0，就不会有寄存。如果 buffering 的值取 1，访问文件时会寄存行。如果将 buffering 的值设为大于 1 的整数，表明了这就是的寄存区的缓冲大小。如果取负值，寄存区的缓冲大小则为系统默认。<br>
encoding: 　　一般使用utf8<br>
errors: 　　报错级别<br>
newline: 　　区分换行符<br>
closefd: 　　传入的file参数类型<br>
opener:　　<br>

不同模式打开文件的mode参数列表：<br>
>"r":　以只读方式打开文件（文件已存在）。文件的指针将会放在文件的开头。这是默认模式。<br>
"rb":　以二进制格式打开一个文件用于只读（文件已存在）。文件指针将会放在文件的开头。<br>
"r+":　打开一个文件（文件已存在）用于读写。文件指针将会放在文件的开头，写为追加在文件末尾写。<br>
"rb+":　以二进制格式打开一个文件用于读写。文件指针将会放在文件的开头。<br>
"w":　打开一个文件只用于写入。如果该文件已存在则打开文件，并从开头开始编辑，即原有内容会被删除。如果该文件不存在，创建新文件。<br>
"wb":　以二进制格式打开一个文件只用于写入。如果该文件已存在则打开文件，并从开头开始编辑，即原有内容会被删除。如果该文件不存在，创建新文件。<br>
"w+":　打开一个文件用于写读。如果该文件已存在则打开文件，并从开头开始编辑，即原有内容会被删除。如果该文件不存在，创建新文件。<br>
"wb+":　以二进制格式打开一个文件用于写读。如果该文件已存在则打开文件，并从开头开始编辑，即原有内容会被删除。如果该文件不存在，创建新文件。<br>
"a":　打开一个文件用于追加写。如果该文件已存在，文件指针将会放在文件的结尾。也就是说，新的内容将会被写入到已有内容之后。如果该文件不存在，创建新文件进行写入。<br>
"ab":　以二进制格式打开一个文件用于追加写。如果该文件已存在，文件指针将会放在文件的结尾。也就是说，新的内容将会被写入到已有内容之后。如果该文件不存在，创建新文件进行写入。<br>
"a+":　打开一个文件用于追加读写。如果该文件已存在，文件指针将会放在文件的结尾。文件打开时会是追加模式。如果该文件不存在，创建新文件用于读写。<br>
"ab+":　以二进制格式打开一个文件用于追加读写。如果该文件已存在，文件指针将会放在文件的结尾。如果该文件不存在，创建新文件用于读写。<br>

#### file 对象<br>
file 对象使用 open 函数来创建<br>
File对象的属性列表：<br>
>file.closed	返回true如果文件已被关闭，否则返回false。<br>
file.mode	返回被打开文件的访问模式。<br>
file.name	返回文件的名称。<br>
file.softspace	如果用print输出后，必须跟一个空格符，则返回false。否则返回true。<br>

file 对象常用的函数：<br>

|序号|函数|用途|
|:---:|:--- |:---|
|1|file.close()|关闭文件。关闭后文件不能再进行读写操作。|
|2|file.flush()|刷新文件内部缓冲，直接把内部缓冲区的数据立刻写入文件, 而不是被动的等待输出缓冲区写入。|
|3|file.fileno()|返回一个整型的文件描述符(file descriptor FD 整型), 可以用在如os模块的read方法等一些底层操作上。|
|4|file.isatty()|如果文件连接到一个终端设备返回 True，否则返回 False。|
|5|file.next()|用于读取文件内容，每次读取一行，其指针位置不发生变化|
|6|file.read([size])|从文件读取指定的字节数，如果未给定或为负则读取所有。<br>size 未指定则返回整个文件，如果文件大小 >2 倍内存则有问题，f.read()读到文件尾时返回""(空字串)。|
|7|file.readline([size])|读取整行，包括 "\n" 字符。|
|8|file.readlines([sizeint])|读取所有行并返回列表，若给定sizeint>0，则是设置一次读多少字节，这是为了减轻读取压力。返回包含size行的列表, size 未指定则返回全部行。|
|9|file.seek(offset[,whence])|作用是：改变当前文件指针的位置。Offset变量表示要移动的字节数,默认是针对于起始位置开始，表示从头测试；whence变量指定开始移动字节的参考位置。|
|10| file.tell()|返回文件当前位置。一个整数,表示当前文件指针的位置(就是到文件头的比特数),换句话说，下一次的读写会发生在文件开头这么多字节之后。|
|11| file.truncate([size])|截取文件，截取的字节通过size指定，默认为当前文件位置。|
|12| file.write(str)|将字符串写入文件，返回的是写入的字符长度。|
|13| file.writelines(sequence)|向文件写入一个序列字符串列表，如果需要换行则要自己加入每行的换行符。|

注：file.seek(offset[,whence]）<br>
1、文本模式下<br>
>whence=0 缺省值,表示从头开始，offset只能接受正整数 <br>
whence=1 表示从当前位置开始，offset 只接受0<br>
whence=2 表示从EOF（表示文件末尾） 开始，offset 只接受0 <br>

2、二进制模式：<br>
>whence=0 缺省值，表示从头开始offset只能是正整数 <br>
whence=1 表示从当前位置开始，offset可正可负 <br>
whence=2 表示从EOF开始，offset可正可负<br>

　　二进模式支持任意起点的偏移，从头、从尾、从中间位置开始向后seek可以超界，但是向前seek的时候，不能超界，否则会报异常。size 表示读取多少个字符或字节，负数或None表示读取到EOF。<br>

#### 编码、换行以及缓冲处理<br>
**编码问题：**<br>
　　Windows的默认编码规则是cp936国标，cp936是GBK 的一种，GBK不是国际标准，GB2312是国际标准。<br>
　　Linux的默认编码规则是utf-8，Unicode传输编码，其一个汉字是3个字节，偶尔可能是4个字节，其总共占6个字节。而Gbk 是 a0 开始的，在utf-8中不能识别。<br>
　　GBK 是顺序的中文编码，GBK是两个字节。utf-8大多是3个字节，可能是4个字节。<br>
1、写入字典文件（包含中文）时，首先把字典转化为json字符串，同时做好编码处理，再写入。<br>
>f.write(json.dumps(dict, ensure_ascii=False))<br>

2、file对象创建时，指定好文件的编码格式：encoding=‘utf-8’<br>

**newline换行问题：**<br>
　　可使用newline进行指定换行符，文本模式中，换行的转换可以为None，"",'\r','\n','\r\n'<br>
　　读取时，None 标识'\r'，'\n'，'\r\n'都被转换为'\n'， ""（空字符串）表示、不会自动转换通用换行符,其他合法字符表示换行符就是指定字符，就会按照自定字符分行，重新定义分割符。<br>
　　写入时，None 表示"\n"都会被替换为系统缺省分割符os.linsep,'\n'或"" 表示"\n"不替换，其他合法字符表示换行符"\n"都会被替换为指定分割符。<br>
　　文件描述符：<br>
　　　　默认的三种，0 标准输入， 1 标准输出， 2 错误输出<br>
　　　　文件描述符可表示打开文件的数量，当打开文件数量达到系统打开文件数量上限时，则会报错<br>

**缓冲处理：**<br>
3、将文件内容写入到硬件设备时，使用系统调用，这类I/O操作的时间很长。为了减少I/O操作的次数，文件通常使用缓冲区。（有足够多的数据才进行系统调用）文件的缓冲行为，分为全缓冲，行缓冲，无缓冲。如果没有缓冲会造成资源的浪费（未写满一个块，依旧写入，使用缓冲，只有当写满之后，才会插入）<br>
解决方案：<br>
>全缓冲：open函数的buffering设置为大于1的整数n，n为缓冲区大小<br>
行缓冲：open函数的buffering设置为1<br>
无缓冲：open函数的buffering设置为0<br>

注意：python3中文本文件不能设置buffering = 0，而二进制文件可以。

#### 文件状态、临时文件以及文件的内存映射<br>
访问文件的状态：<br>
1、文件的类型（普通文件，目录，符号连接，设备文件。。。）<br>
2、文件的访问权限<br>
3、文件的最后的访问/修改/节点状态更改时间<br>
4、普通文件的大小<br>
解决方法:<br>
方法一：使用标准库os模块下的三个系统调用stat，fstat，lstat获取文件状态。<br>
方法二：标准库中os.path下一些函数，使用起来更加简洁<br>

临时文件：<br>
某些项目，我们从传感器采集数据，每收集1G数据后，做数据分析，最终只保存分析结果。这样很大的临时数据如果常驻内存，将消耗大量内存资源，我们可以使用临时文件存储这些临时数据（外部存储）。临时文件不用命名，且关闭后会自动被删除。<br>
解决方案：<br>
可以使用标准库中tempfile下的temporaryFile，NamedTemporaryFile。<br>

文件的内存映射：<br>
1，在访问某些二进制文件时（通常使用read()和write()方法的时候，都是以流的形式读写，一个字节一个字节的顺序进行），希望能把文件映射到内存中，（像数组一样访问）可以实现随机访问。（framebuffer设备文件）<br>
2，某些嵌入式设备，寄存器被编址到内存地址空间，我们可以映射/dev/mem某范围，去访问这些寄存器。<br>
3，如果多个进程映射同一个文件，还能实现进程通信的目的。<br>
解决方案：<br>
使用标准库中mmap()函数，它需要一个打开的文件描述符作为参数。<br>
